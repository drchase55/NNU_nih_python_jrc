# Python is case-sensitive

This means capitalization matters in a name, whether it is a variable, module, method, class or function.<br>
Not all languages are this way, so be mindful of your variables in particular, so you don't create a nasty "bug" or program error that is hard to detect.
<br><br>
For example, "var" and "Var" are not the same reference.
<br><br>
There are naming standards for different languages. Python has one called the <b>PEP 8 Style Guide</b>. You can find it here: https://peps.python.org/pep-0008/
<br>
We will try to stay consistent with the PEP guide in this course. But, when you start writing code, don't worry too much about the particular "style", rather try to be consistent throughout so your readers know what to expect.
<br>
<br>
Try the following code:

```python
# The style guide says variables should be all lowercase with words separated by the underscore
my_var = "Python is case-sensitive"
```

Now see what happens when you "miscase" a variable.

```python
print(my_Var)	# throws an error, the case is different from the declaration
```

# Strong vs Weak vs Dynamic Typing- for programmers

Some languages (like C++) require you to know the type of variable you are creating at the moment of creation.
<br>
<br>
This is what is known as a strongly-typed language
<br>
<br>
The compiler checks your usage to make sure you are passing/using variables of the correct type in the correct way. 
- For instance, if you want to store someone's age in a variable (a numerical value) it will not accept "P" and the compiler can ensure that an error will get thrown if somebody tries to do that.
<br>
<br>
Some languages, like Javascript, don’t care about types. We call them weakly typed. You can get some very weird and difficult to find errors in weakly typed languages so you have to be especially careful.

Python takes the middle road between the two.
<br>
<br>
It uses <b>dynamic typing</b>. This means you don't have to declare the type of variable in advance. Instead Python will determine what "type" the variable is by what information you assign to it.
<br>
<br>
Note that even though we can have dynamic references in Python, it is still considered a strongly-typed language because it doesn’t coerce different types automatically.

Python also allows you to change the variable type at runtime if necessary.
<br>
<br>
Generally it's best to avoid doing this however, as it can lead to difficult to maintain code, but here's how it might work:

```python
# Character type
var1 = "f"
# Change to numeric type by reassigning with a number 
var1 = 7
```

Try the code below to see how it works.

```python
a=10
print(type(a))  #two commands here-- the parenthesis first assesses the type of variable of a, then print, 2nd, to the screen
```

Now a boolean:

```python
a=False    # Note the case is also important for values here, "false" will throw an error
print(type(a))
print(a==True)   #a==True asks Python to evaluate if a is equal to True, not the word but the boolean value. Try changing it to False
```

# Value vs Reference

You’ll hear the terms “pass by value” and “pass by reference." What do they mean?
<br>
<br>
Pass by value means pass a copy of a variable. In computer parlance, you create a new variable with the same contents as the original.
These differences can affect the outcome of your programs, so take a moment to consider the differences.

Here’s why it’s important. Consider a variable passed by *value*:


```python
# passing a variable by value
var1 = 1
var2 = var1
```

Any change made to var2 will not affect var1. Check the values of both variables:

```python
# Change one of the variables
var1 = 2

# Compare the two variables
print("var1")
print(var1)
print("var2")
print(var2)   #var2 was assigned to the value of variable 1 before it got changed to =2
```

Now consider a variable passed by *reference*:


```python
# Pass by Reference
# Basic types are passed by value.
# Complex types, like a list, are passed by reference.
# Don't worry about the actual code, we will cover the list object and range function in the next section.
print("var2:", var2)
var1 = list(range(3))
# now assign the variable
var2 = var1
# check the new variable against the old variable
print("var1")
print(var1)
print("var2")
print(var2)
```

Any change made to var1 will affect var2 (let’s try it) because var1 is a more complex type

```python
# now change the old variable
var1.append(10)

# check the new variable against the old variable
print("var1")
print(var1)
print("var2")
print(var2)
```

## Capturing User Input

On many occasions you will want to ask your user for input from the keyboard, which becomes the value for a variable
<br>
<br>
We can prompt a user for input, then use it in the program, like this:

```python
# User Input: Note the input comes back as a string.
player1 = input("What is your name, player? ")
print("Welcome, ", player1)

# Square a Number. Again, even though we are asking for a number,
# the input will still be a string so we will have to convert
# using what is called a "cast" (see below).
# In this case, we are casting the string to a float (floating point number).
num1 = input("Enter the number you would like me to square: ")
num2 = float(num1)
print("The result is ", num2 * num2)
```

You can *cast* a value to change its type.
<br>
<br>
Casting is useful when working with variables/values of different types.
- For example you can cast strings to numbers, using int() or float():
- Note that leading and trailing whitespace with numbers inside will convert OK.

```python
# Type Conversion with trailing or leading spaces is OK.
print(int("  45"))
print(float("4.769  "))
```

The case has to make logical sense of the interpreter will throw an error:

```python
# This conversion doesn't make sense and will throw an error.
int("b6")
```

Note that internal whitespace in a cast like this will also cause an error:

```python
# This conversion contains spaces embedded between numbers and will also throw an error.
float("4.68 97")
```
