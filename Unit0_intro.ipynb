# Getting Started with Github and Jupyter Notebooks

## Welcome!

The National Institutes of Health (NIH) has multiple cloud computing programs with goals to improve access to data, modernize biomedical research, and promote equity among institutions
<p style="color:red;font-family:arial"> However... </p> 
Many biologists and students who could <i>benefit</i> from modern programming and cloud-computing tools do not have enough background to take advantage of the resources that the NIH has commissioned and posted so that you can freely download the materials from Github ("a web-based platform that allows users to share, store and collaborate on code"). Of course, that requires that you already know what to do to access and use those materials. We would like to help you to overcome this hurdle.

## Prerequisites

<p style="background:blue;color:white;font-family:arial">This unit assumes that you have no experience with GitHub, Python, or even Jupyter notebooks (even though you are using a Jupyter notebook right now).</p>

## Objectives

The goal of this module is for you to become comfortable with the following:
* Navigating the various material sources within GitHub repositories
* Downloading materials from the NIH "Sandbox" (source code, bioinformatics, and datasets)
* Creating an account on GitHub
* Uploading materials (e.g., your own data for storage & FAIR sharing) to a GitHub repository
* Downloading and installing Python or other similar thingies so you can run other Jupyter-notebook based tools
* <u>Run</u> a Python program to return the complement of a DNA sequence, to demonstrate Python.
