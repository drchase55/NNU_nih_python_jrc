# Strong vs Weak vs Dynamic Typing

Now back to variables.

<br>
<br>
This is what is known as a strongly-typed language

The compiler checks your usage to make sure you are passing/using variables of the correct type.

Python takes the middle road between the two.

# Value vs Reference

You’ll hear the terms “pass by value” and “pass by reference” quite a bit. What do they mean?
<br>
<br>
Pass by value means pass a copy of a variable. In computer parlance, you create a new variable with the same contents as the original.


Here’s why it’s important. Consider a variable passed by *value*:


```python
# passing a variable by value
var1 = 1
var2 = var1
```

Any change made to var2 will not affect var1. Check the values of both variables:

```python
# Change one of the variables
var1 = 2

# Compare the two variables
print("var1")
print(var1)
print("var2")
print(var2)
```

Now consider a variable passed by *reference*:


```python
# Pass by Reference
# Basic types are passed by value.
# Complex types, like a list, are passed by reference.
# Don't worry about the actual code, we will cover the list object and range function in the next section.
var1 = list(range(3))
# now assign the variable
var2 = var1
# check the new variable against the old variable
print("var1")
print(var1)
print("var2")
print(var2)
```

Any change made to var1 will affect var2 (let’s try it):

```python
# now change the old variable
var1.append(10)

# check the new variable against the old variable
print("var1")
print(var1)
print("var2")
print(var2)
```

# String Formatting

Python provides some useful formatting options to make it easier to work with strings:
<br>
<br>
(For more detail, look here: https://docs.python.org/3/library/string.html#string-formatting)
<br>
<br>
There is the **.format** method(). 
<br>
<br>
You provide a string and insert a dynamic (i.e. from a variable) value
using the curly braces, like this:

```python
print("Hello, my name is {name}".format(name="Rufus"))
```

You can specify more than one dynamic value.

```python
print("Hello, my name is {fname} 	{lname}".format(fname="Rufus",lname="McGee"))
```

You can also use indexers.

```python
print("Hello, my name is {0} 	{1}".format("Rufus","McGee"))
```

Or even use empty curly braces to indicate you will provide the proper order yourself.

```python
print("Hello, my name is {} {}".format("Rufus","McGee"))
```

Finally you can even add some additional string formatting options like extra spaces or tabs to align your output.

```python
print("Hello, my name is {0:_<5} {1}".format("Rufus","McGee"))
```

# Dates & Times

Dates and times are a bother in most languages, including Python.
- This is because dates are stored internally as numbers, but usually expressed as a string, like "July 4, 1776."

Fortunately, Python includes a **datetime** module to help with this.
<br>
<br>
**datetime** provides the **datetime** class as well as a number of method and functions to help a programmer work with translating dates back and forth and performing mathematical operations on them.
<br>
<br>
We'll do a deep dive into classes in later modules. For now, understand a class is a kind of super type of variable used in object-oriented languages like Python.
<br>
<br>
Because **datetime** is a class it contains specialized methods and properties specific to dates and time. You can use them to help you manage the underlying datetime values as well as present them in a nice, readable form to your users.

```python
import datetime

x = datetime.datetime.now()
print(x)

print(x.year)
print(x.day)
print(x.strftime("%A"))

# first day of 2023
start_date = datetime.date(2023, 1, 1)
print(start_date.strftime("%A"))
```
