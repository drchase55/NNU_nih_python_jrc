### Reading a file

To access a file in Python or most languages, you need to create a *file handle*.
- Do this using the **open** function, with a "flag" telling the operating system (OS) whether the file is open for read-only or read/write operations.
    - The flag is often referred to as the mode.
    - The difference between read and read/write modes is the type of lock the OS will apply. Read-only files can be shared among processes while read/write locks are exclusive.

```python
# Open a text file for reading. countries is the handle.
countries = open("." + os.sep + "Datasets" + os.sep + "countries.txt", "r")
                 
# Print out the results of the entire file, line by line.
for line in countries:
    print(line)

# Close the file when done. This releases the handle.
countries.close()
```

## with statement

Just like with memory, developers have a nasty habit of not cleaning up their file handles.
- Having an open file handle means the file won't be available for other applications.
- Sometimes it gets so bad you have to reboot the computer to force the OS to let go of the files. That's not good.

Modern languages provide helper commands to automatically handle cleanup of things like memory and file handles which can get missed by developers.
- Python provides the **with** statement for this purpose. 

Use **with** to ensure that open handles are closed and destroyed properly.
- The also works with database connections, as we will see later.​

Note that objects declared using **with** exist for the code block only​.

```python
# Use with to open files to ensure closure.
file_path = "." + os.sep + "Datasets" + os.sep + "green_eggs_ham.txt"

# Note that we didn't specify the type of lock we wanted. The default is "r" or read-only.
with open(file_path) as f1: 
    dat = f1.read()

print(dat)
```

### CSV Files

CSV files are special in that they are a common exchange format for datasets​.

CSV stands for comma separated values.​

  - They may or may not contain headers one the first line.​

  - Records are return separated​.

Python provides a **csv** library with methods for reading and writing csv files.​

CSV files are just text, but they are formatted.​

So the process is much the same as before except you can get back formatted data in a form such as a list of lists.​

```python
# Example using CSV files with the csv module.
import csv

file_path = "." + os.sep + "Datasets" + os.sep + "states.csv"
f_handle=open(file_path,"r")

# Create csv_reader out of csv file.
csv_reader=csv.reader(f_handle) # Returns csv reader object
print(type(csv_reader))

myData=list(csv_reader)
print(myData) # List of list returning what looks like a table.
for myRow in myData: # Iterating each element as list.
    for myCol in myRow:# Iterating each element of list.
        print(myCol,"\t",end="")
    print()
```

### Zip files

Zip files are compressed files containing one or more internal files​.

The compression is “lossless” which means you get an exact copy of the original file(s) that were compressed​ after decompression.

Python includes a library called Zipfile for handling *zip* (compress) and *unzip* (decompress) operations​.

Creating a zip file:
- Use ZipFile to create an empty zip​.
- Add files as you go​.
- Close the file​.

Let's see how this works:

```python
# Creating a zip file from a couple of regular files.
# In this case we are preserving directory information.
from zipfile import *
file_path = "." + os.sep + "Datasets" + os.sep
myfile1=ZipFile(file_path + ".." + os.sep + "SupportFiles" + os.sep + "test_zip.zip","w", ZIP_DEFLATED)
myfile1.write(file_path + "titanic_train.csv")
myfile1.write(file_path + "titanic_test.csv")
myfile1.close()
print("test_zip.zip file is created successfully")
```

```python
# Creating a zip file from a couple of regular files.
# In this case we are NOT preserving directory information.
import zipfile

file_path = "." + os.sep + "Datasets" + os.sep
with zipfile.ZipFile(file_path + ".." + os.sep + "SupportFiles" + os.sep + "test_zip.zip","w") as zipf:
    zipf.write(file_path + "titanic_train.csv", arcname="titanic_train.csv")
    zipf.write(file_path + "titanic_test.csv", arcname="titanic_test.csv")

print("test_zip.zip file is created successfully")
```

### Working with zip files

Use **is_zipfile()** to test if the file is a zip​.

Use **printdir()** to print out the list of files in a zip​.

Use **namelist()** to pull the names and extract individual files from the zip​.

```python
import os
from zipfile import ZipFile, ZIP_STORED

file_path = "." + os.sep + "SupportFiles" + os.sep + "test_zip.zip"
# Crack open the zip file
my_file1=ZipFile(file_path,"r", ZIP_STORED)

# Pull a list of filenames from the zip.
my_filename=my_file1.namelist()
# Iterate of the list of names.
for my_file in my_filename:
    print("File Name is : ", my_file)
    # Dump the contents of the file.
    with my_file1.open(my_file) as file:
        contents = file.read().decode("utf-8") # Assuming file is in UTF-8 format.
        print("The Contents of this file are:\n")
        print(contents)
```

You may need to save or *serialize* Python objects to disk and restore or *deserialize* them from disk.​

In true Python fashion, serialization is called **pickling** and deserialization is called **unpickling​**.

Pickle objects if text files are either too large, too complex, or not desired​ for your purposes.
- Pickle files are known as semi-structured in that they contain meta-data along with the data so they can be restored without knowing the internal structure of the data beforehand.

To pickle an object to disk, use the **dump()** method.

```python
# Pickle an object.
# First let's define a simple object for pickling.
import os
file_path = "." + os.sep + "SupportFiles" + os.sep + "student.dat"
if os.path.exists(file_path):
    os.remove(file_path)

import pickle

class Student:
    def __init__(self, name, age, roll_num, home_town):
        self.name = name
        self.age = age
        self.roll_num = roll_num
        self.home_town = home_town

    def my_display(self):
        print(self.name, "\t", self.age, "\t", self.roll_num
              , "\t", self.home_town)
        
# Pickle files are binary files.
# Create a new pickle file and serialize the Student object.
with open(file_path, "wb") as pickle_file:
    student = Student("Rufus", 21, 101, "Nampa")
    pickle.dump(student, pickle_file)
    print("Pickling of Student object completed.")
```

To unpickle, use the **load()** method​.

```python
# Verify that the object was pickled correctly by loading it back
with open(file_path, "rb") as pickle_file:
    loaded_student = pickle.load(pickle_file)
    print("Loaded Student object from pickle file:")
    print("Printing Student Information after unpickling")
    loaded_student.my_display()
```

Another popular format used for reading and writing files is JSON.​

- Less verbose but just as flexible as XML, it is the preferred format for transmitting text data over the internet​.

- Python provides a json library to read/write JSON files​.

```python
# Example of working with JSON.
import json, os

# A simple list for serialization/deserialization.
out_list = [1,"the",4]

# Write JSON out to a file.
file_path = "." + os.sep + "SupportFiles" + os.sep + "test.json"
with open(file_path, "w") as output_file: 
    json.dump(out_list, output_file)

## Load the results back from JSON
with open(file_path,"r") as input_file:
    in_list = json.load(input_file)

# Compare the two lists. True if they are the same, False otherwise.
in_list == out_list
```

## Web Services

There are multiple ways to handle web services. We will focus on the most common one, using a REST API​.

Python supplies the requests module to work with REST calls​.

You will need the following:

  - A URL​

  - A **requests** object​

The web service will send back a response.​

There are a number of request “types” you can send to a web service​.

Each has a specific function:​

>Get: Retrieve data​.
>
>Post: Create a new resource​.
>
>Put: Update an existing resource​.
>
>Patch: Partially update an existing resource​.
>
>Delete: Delete a resource​.

A request is given a response. That response will include a number of items, including a status code indicating how the request was handled​.

Check these codes to see whether or not a request succeeded​.

>200 – Success​
>
>400 – Bad request​
>
>404 – Resource not found​
>
>500 – Server error​
>
>Many others​

For this to work, the web service needs to establish “endpoints.”​

Endpoints are simply a URL that is used to access an API call​.
- Similar to a method call of a class object​.

Use a get request to acquire a resource (usually data)​.

There is a URL we can use for testing​:

```python
# Python example for interacting with the web service.
import requests
# This is a test url you can use to try out your code.
api_url = "https://jsonplaceholder.typicode.com/todos"
# Create a "todo" object to post to the endpoint.
todo = {"userId": 1, "title": "Buy milk", "completed": False}
response = requests.post(api_url, json=todo)

# Check to see if it was OK
# A 201 status code means the post was successful and the resource was created.
print(response.status_code)
response.json()
```

Use **post** to create a resource you or someone else will want to use later​:

  - Usually in the form of a database record​.

```python
import requests
# Example post command.
api_url = "https://jsonplaceholder.typicode.com/todos"
todo = {"userId": 1, "title": "Buy milk", "completed": False}
response = requests.post(api_url, json=todo)
print(response.status_code)
response.json()
```

Use **put** to update existing data​:

  - The data should already be in the database, otherwise you would use a post command​.

  - We believe, however, a number of services will accept a post command too​.

```python
# Example put command - update the list item.
todo = {"userId": 1, "title": "Wash car", "completed": True}
response = requests.put(api_url, json=todo)
print(response.status_code)
response.json()
```

If your REST API supports it, you can submit a partial update using the **patch** command​:

  - Sometimes the record is rather large so it would be nice to not have to pass the entire thing​ all at once.

```python
# An example using the patch command.
import requests
api_url = "https://jsonplaceholder.typicode.com/todos/10"
todo = {"title": "Mow lawn"}
response = requests.patch(api_url, json=todo)
print(response.json())
print(response.status_code)
```

Of course, what you can add, you should also be able to delete​

```python
# Example requesting a delete from a web service.
import requests
api_url = "https://jsonplaceholder.typicode.com/todos/10"
response = requests.delete(api_url)
print(response.status_code)
print(response.json())  # Should be empty.
```
