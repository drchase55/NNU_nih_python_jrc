## Static Variables 

To this point we have dealt with a few types of variables, which were all meant to operate with a certain section of code, or, if they were global, everywhere.

There is a special kind of varible, however, that is different from those but very useful in situations where a value needs to be "shared".

Imagine you are at the DMV waiting to get your license renewed. You pull a ticket from the dispenser. When do you know it is your turn? How can you tell how many poor souls are in front of you? There needs to be a display that everyone can see that shows the most current ticket being served, otherwise people have no idea about when their number is next or even close to next.

Static variables are shared between instances of a class. Think of each DMV ticket as an instance variable. Everybody gets one of their own. However, the display of the current ticket is like a static variable. Everyone can see it and when you get to the counter, you get to update the value on the display.

With that picture in your mind, let's formalize our discussion of static variables. They are declared inside a class, but NOT in the constructor or any other class method unless the class name is used.
- This means they exist outside any individual object created and are attached, in a sense, to the definition itself.
  - Hence, when accessing a static variable, it's not as an attribute of the object, rather it is an attribute of the class itself. 
- Note that the instance variables we have used in the past all required the use of the **self** keyword.

In other languages they are also known as static variables but usually declared with the **static** keyword.​

In Python they are often referred to as object or class variables, but we'll stick with the more popular vernacular "static."

Let's see some examples:

```python
# Create a class with a static variable.
# Note the static variable is declared inside the class but outside any method.
class Student:
    universityName = "NNU" # static variable

    def __init__(self, myName, gradeLevel):
        self.myName = myName
        self.gradeLevel = gradeLevel
```

One convention Python allows is to declare a static variable within a class constructor using the class name.​

```python
# Create a class with a static variable.
# This nice thing about this approach is that the code is more explicit.
# We "know" we are creating a static variable, rather than just assuming it.
class Student:
    
    def __init__(self, myName, gradeLevel):
        self.myName = myName
        self.gradeLevel = gradeLevel
        Student.universityName = "NNU"  # static variable
```

It doesn't have to be the constructor. You can also declare a static variable within an instance method, using the class name​:

```python
# Create a class with a static variable.
# You might want to declare it this way if it improves the readibility of your code
# or makes it more explicit as to what the variable is intended to do.
class Student:
    
    def __init__(self, myName, gradeLevel):
        self.myName = myName
        self.gradeLevel = gradeLevel
        
    def set_NNU(self):
        Student.universityName = "NNU"  # static variable
```

Yet another way is to declare within a **class method** using the class name. We'll get more into class methods in a moment.​

```python
# Create a class with a static variable.
# We declare a class method, and create the variable inside.
class Student:
    
    def __init__(self, myName, gradeLevel):
        self.myName = myName
        self.gradeLevel = gradeLevel
        
    @classmethod
    def set_NNU(cls):
        cls.universityName = "NNU"  # static variable
```

Declare a static variable within a static method using the class name​.

```python
# Create a class with a static variable.
# We declare a class method, and create the variable inside.
class Student:
    
    def __init__(self, myName, gradeLevel):
        self.myName = myName
        self.gradeLevel = gradeLevel
        
    @staticmethod
    def set_NNU():
        Student.universityName = "NNU"  # static variable
```

Usage of a static variable is the same as any other type of variable, except changes to that variable in any one instance affect all other instances of that class.​

```python
class Student:
    universityName = "NNU" # class variable
    def __init__(self, myName, gradeLevel):
        self.myName = myName
        self.gradeLevel = gradeLevel

student1 = Student("Rufus McGee", "Sophomore")
student2 = Student("Alice McGee", "Junior")
print(student1.universityName)
print(student2.universityName)
Student.universityName = "Northwest Nazarene University"
print(student1.universityName)
print(student2.universityName)
```

### Static Methods vs. Class Methods

Just to clear up any confusion here about the difference between static methods and class methods, let's quickly discuss them.

How they are the same:
- They are both decorators.
- They both define special types of methods within a class.
- Both define methods called on the class rather than on the instance.
- They can both modify class state.

How they are different:
- A class method operates on the class using the cls variable. 
- A static method operates on the class using the class name.

So why do we care?
- Use static methods as a way to organize related functions (such as math functions) that aren't really related to any state information.
- Use class methods for operations on state variables.

### Modifying Static Variables

Python provides lots of ways to make changes to static variables.

  - Are you surprised? Are you also surprised there is no “right” way. It depends upon a given situation which is best​.

Here are the available ways to modify a static variable:
- Modify inside the class using the class name​.
- Modify inside the class using the cls variable​.
- Modify using a static method​.

```python
# Demonstrate all the ways one might modify a static variable.
class Student:
    # Declare the class/object/static variable
    university_name = "Northwest Nazarene University"  
    
    def __init__(self, my_name, grade_level):
        self.my_name = my_name
        self.grade_level = grade_level
        
    # Modify the static variable using the cls variable.
    @classmethod
    def set_NNU1(cls, university_name):
        cls.university_name = university_name

    # Modify the static variable using the class name.
    @classmethod
    def set_NNU2(cls, university_name):
        cls.university_name = university_name

    # Modify the static variable using a static method.
    # Note that we won't normally do this, but showing this just for comparison.
    @staticmethod
    def set_NNU3(university_name):
        Student.university_name = university_name
```

Modifications are initiated via method calls either from the instance or the overall class (for a static method)​.

```python
# Instantiate the class.
student1 = Student("Rufus McGee", "Sophomore")
# Modify using cls variable
student1.set_NNU1("NNU")
# Modify using the class name.
student1.set_NNU2("NNU")
# Modify using a static method.
Student.set_NNU3("NNU")
```

### Class Methods 

For the most part, you will use class methods when working with class variables.

You declare them explicitly using the **@classmethod** decorator​.

The first parameter is the **cls** variable, which is used to access the current class object​.
- Similar to **this** in other languages.

Class methods are useful in exposing instance variables.​

```python
class Major:
    myConcentration = "Data Science" # class variable
    def __init__(self, my_name):
        self.my_name = my_name
    
    @classmethod # decorator
    def Concentration(cls, degree): # class method with parameter
        # accessing class variable inside class method
        print(f"{degree} has a concentration in {cls.myConcentration}") 

Major.Concentration("Computer Science")
```

## Inner Classes

You might occasionally have a need for a dependent class of another class​.

  - This is not inheritance. What we mean by dependence here is a class that requires another class to complete some sort of function.​
  - Like a university needs students​. Not a lot can happen without them.

Python allows us to declare classes within classes to handle this particular scenario.​

In this case, the outer class is created first, followed by the inner class or classes:​

```python
# Class within a class
# Outer class
class University:
    def __init__(self, my_name):
        self.my_name = my_name
    
    def my_display(self):
        print(self.my_name)
    
    # Inner class
    class Info:
        def __init__(self):
            pass # Useful code goes here.            
        
        def my_display1(self):
            print("Inner Class")

# Access the inner class using dot notation
university = University("NNU")
# The inner class doesn't get created automatically, so let's do it
u_info = university.Info()
u_info.my_display1()
```

Now, if you did want the inner class to be created automatically, you need to create a class or instance variable inside the outer class and create it in the constructor of the outer class. Like this:

```python
# Class within a class #2, auto-creates the inner class
# Outer class
class University:
    # Constructor
    def __init__(self, my_name):
        self.my_name = my_name
        # Create the inner class
        self.inner_info = self.Info()
        
    def my_display(self):
        print(self.my_name)
    
    # Inner class
    class Info:
        def __init__(self):
            pass # Useful code goes here.            
        
        def my_display1(self):
            print("Inner Class")

# Access the inner class using dot notation
university = University("NNU")
# The inner class gets created automatically, so no need to do it here
university.inner_info.my_display1()
```

## Garbage Collection

Creating objects requires allocating a lot of memory.
- In the distant past, programmers were responsible for their own garbage collection, which is really just memory recycling.
- Needless to say, it didn't always go very well.
- "Garbage collection" is a key component of an object-oriented framework. It tries to maintain sufficient memory for applications without putting undue burdens on the processor. It does this by recycling memory only when it is needed or when the CPU is sufficiently idle that performance won't be affected.​

Python provides a “garbage collector” that runs in the background, cleaning up no-longer used memory​ as appropriate.

By default, garbage collection is enabled, but it can be disabled for performance reasons​.

In case you need to manually clean up an object, you can use the object’s *destructor*. This is a method that every object has that gets called when the object is in the process of going out of scope and needs to be cleaned up. You can create custom code to perform specialized cleanup operations.

Most of the time you will use the default destructor, but if you want to create your own you override the method **\__del__**

```python
# Example of a class with a custom destructor.
class Demo:
    def __init__(self):
        print("Initializing the object")
    def __del__(self):
        print("clean up activities are performed...")
```

## Polymorphism

Polymorphism is another important facet of object-oriented programming.
- It is about changing behaviors depending upon the situation​.

  - For example, + adds two numbers but concatenates two strings​

  - Another example is you “eat” a steak by chewing it, but you “eat” soup by sipping it​.

It makes code easier to understand and maintain​ (because the function "eat" is always eat).

To do this, Python uses a method called “Duck Typing”​

  - Instead of checking an object's type it assumes the necessary behaviors will work as anticipated at runtime.
  - In essence, instead of determining whether the object "is a duck" you check whether it behaves in some fashion "like a duck".
  - This also means you could use similar objects, like a bird or goose or chicken so long as they show the same behaviors as the "duck." 
  
Here are the basic steps to program using polymorphism:

- Create a class with a common method​.

- Instantiate the class​.

- Pass the class and call the method​.

```python
# Note this code is just for illustration. It won't run because we haven't defined the classes.
# Consider three different animals.
duck = Duck()
bird = Bird()
dog = Dog()

# Now consider a common behavior. Each object has a different way of making noise.
# But they all "make noise". This is polymorphism.
make_noise(duck)
make_noise(bird)
make_noise(dog)
```

But what if the object doesn’t have the right method (or other component)?​

If this is a concern use the **hasattr(the)** function​

  - Otherwise you will need to trap the AttributeError exception​

```python
def myfunc(object):
    # Pythonic code: Test for the presence of a "talk" and/or "eat" behavior.
    if hasattr(object, 'talk'):
        object.talk()
    elif hasattr(object, 'eat'):
        object.eat()
```

### Overloading

We’ve already seen how operators like + work differently depending upon the operands​ string and number.
- This is an example of operator overloading​.

\+ is tied to the magic method **\__add__** so if we want to extend the operator even more we can modify that method in our class​.
- Overload the + operator by substituting the default **\__add__** with your own version​:

```python
# An example of + overloading.
class MyCylinder:
    def __init__(self, my_weight):
        self.my_weight = my_weight
        
    # overload + operator here
    def __add__(self, other):
        return self.my_weight + other.my_weight
```

### Overriding Class Methods

When inheriting from a class, you will often want to override certain behaviors.
- To do this, you must override the desired method of the parent class.

Python makes it easy to override a class method​.
- Simply re-declare the method in the child class​ and add your custom code.
- Remember, if you still want the functionality of the base class, use the super() method and call the parent class method inside your child method.

Here's how it might work:​

```python
# The parent class subtracts two numbers.
class Parent:
    def result(self, num1, num2):
        print(f"Subtraction of numbers {num1} and {num2} is", num1 - num2)
    
# The child class overrides the parent class to add two numbers.
class Child(Parent):
    def my_display(self):
        print('Hey I am a child class')
    
    def result(self, num1, num2):
        # You can use the super() call to bring in the functionality of the parent.
        print(f"Addition of numbers {num1} and {num2} is", num1+num2)
```

### Abstract Classes

Abstract Classes are a very powerful aspect of Object-Oriented Programming.
- They allow you to define very specific interfaces for behaviors and attributes of a class witout having to implement all the underlying code.
- Think of it like a scale model of an airplane but not the actual airplane.

Python doesn't have native support for abstract classes, but does provide the **abc** module that allows you to create them.​

Abstract classes have some peculiar properties you need to be aware of:
- They must be inheritied from and any non-implemented components coded.
  - In other words, you cannot directly instantiate an abstract class​. You must inherit from the abstract class and create an object from the child class.

#### If you must inherit an abstract class, why bother with them at all?  

  - Why use abstact classes then? Because they can serve as a common object type in your code.
  - But what does that mean?

Abstract classes are known as meta classes because they usually have definitions but not implementations​. What makes them useful is they can serve as a "known/unknown" in your code.

Let's say you want to create a "zoo." A zoo has lots of different kinds of "animals". What do all animals have in common? You might say:
  - All animals need to eat (food).
  - All animals need a place to rest (a cage).

So anytime you add an "animal" to your zoo, you know you need, at a minimum, a cage and food to feed it. The proper food and proper cage will depend upon the animal, but generically you know that an animal must have them in order to be added to your zoo.

Now suppose you have a zookeeper for the zoo. At dinnertime, you wouldn't say "Go feed grass to the zebras in the outdoor enclosure and meat to the lions in the den and fish to the seals in the tank and...". Instead you would say, "Go feed the animals." and let the zookeeper figure out what food goes with what animal and where.

In code, your zoo would be an array of "animals." At suppertime, your zookeeper iterates over that array and determines the food for each and what cage to put the food in, depending upon the individual type. Because of this, the zoo, as a whole, is easier to manage. Now a visitor to the "zoo" comes there because she just wants to see the "animals" and be amazed by their splendor and variety, not what they eat or where they live.

Adding abstract classes in Python requires the following steps:

- Import ABC, abstractmethod​

- Inherit from ABC​

- Use the @abstractmethod decorator for abstract methods​

```python
# Create an abstract class using the abc module.
from abc import ABC, abstractmethod

class Parent(ABC):
    @abstractmethod
    # Abstract methods are placeholders you must complete in the descendant class
    def mymethod1(self):
        pass

    # Concrete methods can be used as it or overridden
    def mymethod2(self):
        print('I am a Concrete method')
```

Note what happens when you try to create an instance of an abstract class:

```python
# This will generate an exception
person = Parent()
```

Now see what happens you don't properly create the base class:

```python
# An example of an abstract class and a malformed child class.
from abc import ABC, abstractmethod
class Person(ABC):
    @abstractmethod
    def my_method1(self):
        pass
    
    @abstractmethod
    def forgot_this(self):
        pass

    def my_method2(self):
        print('I am a Concrete method')

class Student(Person):
    def my_method1(self):
        print("I am overriding abstract method")

# This will throw an exception because we forgot to override forgot_this
myobj = Student()
myobj.my_method1()
myobj.my_method2()
```

Make sure to inherit from the abstract class, then override any and all abstract methods​. Let's build a tiny zoo.

```python
# Create an abstract class Animal
from abc import ABC, abstractmethod
class Animal(ABC):
    def __init__(self):
        self.my_type = "Animal"

    @abstractmethod
    # Abstract methods are placeholders you must complete in the descendant class
    def eat(self):
        pass
    # Concrete methods can be used as it or overridden
    def make_sound(self):
        print("Roar!")

# Inherit from abstract class Animal
class Lion(Animal):
    def __init__(self):
        self.my_type = "Lion"

    def eat(self):
        print("Yummy steak!")

class Zebra(Animal):
    def __init__(self):
        self.my_type = "Zebra"

    def eat(self):
        print("I like grasses and leaves.")
    
    def make_sound(self):
        print("Bray!")
```

```python
# Add our animals to our zoo.
zoo = list()
zoo.append(Lion())
zoo.append(Zebra())

# Let them do their thing.
for animal in zoo:
    print("I am a " + animal.my_type)
    animal.eat()
    animal.make_sound()
```

### Interfaces

Similar to abstract classes is another object-oriented concept called interfaces.
- Object-oriented languages use interfaces to establish an object’s behaviors, regardless of the kind of object involved​
- An interface can establish a "common behavior” among what could be very unlike things.

E.g. All living things “eat”, “breathe.” (This category includes plants as well as humans.)
- So if we want to create a “living thing” it must have a method for “eat” and “breathe.”
- This ensures our "thing" has all the necessary behaviors to be "living."
- It also means any consumer can count on those things being available at runtime.​

Python does not support interfaces, yet, but Python simulates interfaces via abstract classes​.

- Simply create an abstract class without any actual methods​

- Any real (that is a class you can instantiate) class must implement all those methods​

- Any class not implementing all those methods must be another abstract class​

So implementing an “interface” means creating a base, empty method in an abstract class​

  - It must then be overridden in order to work​

Note that in other languages the compiler will force you to implement the method, but Python is OK with it being empty​

```python
# Creating an interface.
# Use an abstract class when you have many behaviors/attributes in common among related objects.
# Use an interface when common behaviors/attributes are few and/or the objects are very different.
from abc import ABC, abstractmethod
class Myengineer(ABC):
    @abstractmethod
    def my_branch(self):
        pass

class MyEEE(Myengineer):
    def my_branch(self):
        print("I am an electrical and electronics Engineer")

rufus = MyEEE()
rufus.my_branch()
```

```python
# Solution
from abc import ABC, abstractmethod
def dog_noise(method):
    def inner(refname):
        if refname.breed.lower() == "dachshund" or refname.breed.lower() == "chihuahau":
            return "Yip!"
        else:
            return method(refname)
    return inner

class Animal(ABC):
    @abstractmethod
    def eat(self):
        pass
    
    @abstractmethod
    def move(self):
        pass
    
    @abstractmethod
    def make_noise(self):
        pass

class Dog(Animal):
    ''' Dogs are awesome'''
    kennel = "Pretty Puppy"
    owner = "Rufus McGee"
    # constructor
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.set_age(age)
        
    @dog_noise
    def dog_bark(self):
        print("In DogBark")
        return "Bark!"
    
    def set_age(self, age):
        if age <= 0:
            raise ValueError('The age must be positive')
        self._age = age

    def get_age(self):
        return self._age
    
    age = property(fget=get_age, fset=set_age)
    
    @staticmethod
    def set_kennel(kennelName):
        Dog.kennel = kennelName
    
    @classmethod
    def set_owner(cls, ownerName):
        cls.owner = ownerName 

    def eat(self):
        print("Dog food! Yum!")
    
    def move(self):
        print("Scamper! Sacmper!")
    
    def make_noise(self):
        return self.dog_bark()

Fido = Dog("Fido", "Beagle", 3)
print(f"{Fido.name}'s owner is {Dog.owner} and he comes from {Dog.kennel} kennel")
Fido.set_owner("Alice Marigold")
Dog.set_kennel("Delightful Dogs")
print(f"{Fido.name}'s owner is {Dog.owner} and he comes from {Dog.kennel} kennel")
print(Fido.dog_bark())

Fido.eat()
Fido.move()
print(Fido.make_noise())
```
