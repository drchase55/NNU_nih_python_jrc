# Graphical User Interfaces in Python (using tkinter)

**Graphical User Interfaces** (GUIs) are a common part of most user programs these days.​

  - They make programs easier to use and understand​.

  - They allow for a lot more functionality​.

  - They are “cooler” and more visually appealing​.

Python provides several libraries for creating GUIs for your programs​. We will use **tkinter**, which is packaged with Anaconda​.

Graphical User Interfaces consist of two major components​:

  - Windows​

  - Widgets​


The *window* is the main interface for the user​.

  - It handles the menus and base visualizations​.

Within the window are *widgets*.​

  - These would be the buttons, views and other controls the user has to create content and interact with the application​.

## Tkinter library in python

Tkinter is a powerful library in Python for creating Graphical User Interfaces (GUIs). It serves as a Python wrapper for the Tk GUI toolkit, and understanding how it "thinks" can help you design and manage GUIs effectively. This module will help you to understand at to practice these design elements. 

### Event-Driven Programming
<u>How tkinter thinks about this:</u> Tkinter operates in an **event-driven** programming model. This means that instead of running sequentially like a typical script, Tkinter waits for events (e.g., a button click, a key press) and responds to them.

**What this means for you:**
- You define the layout and widgets.
- You set up "event bindings" to tell Tkinter what to do when specific events occur (e.g., clicking a button triggers a function).

### Hierarchical Widget Tree (Parent-Child Relationship)
<u>How tkinter thinks about this:</u> Every widget (e.g., buttons, labels, text boxes) in Tkinter is part of a hierarchical tree. At the root of this tree is the main application window (created by Tk()).
- Widgets (e.g., buttons, labels) are children of the root window or other widgets.
- You use layout managers (pack, grid, or place) to arrange widgets within their parent.

**What this means for you:**
When you create a widget, you must specify its parent (e.g., the main window or a frame).

Widgets inherit properties (like size or visibility) from their parent.

### The Mainloop: Tkinter's Brain
<u>How tkinter thinks about this:</u> Tkinter relies on a main event loop, started with **mainloop()**, to keep the application running and responsive. The mainloop continuously listens for events (like button clicks or key presses) and handles them.
**What this means for you:**
- You set up the GUI (create widgets, configure layouts, bind events).
- Once everything is ready, you start mainloop() and let Tkinter take control of the program flow.
- While in mainloop, your program waits for user input and responds as needed.

### Widgets are Objects
<u>How tkinter thinks about this:</u> Widgets in Tkinter (e.g., Button, Label, Entry) are **Python objects.** Each widget has properties (e.g., size, color, text) and methods (e.g., .pack(), .grid(), .config()).

**What this means for you:**
You can create and manipulate widgets programmatically, updating their properties or behavior dynamically (e.g., change the text on a label after clicking a button).

### The Role of Layout Managers
<u>How tkinter thinks about this:</u> Tkinter uses layout managers to control how widgets are displayed inside their parent containers. 

The three main layout managers are:
- pack: Places widgets relative to each other (e.g., top, bottom, left, right).
- grid: Organizes widgets in a grid-like structure with rows and columns.
- place: Allows for absolute positioning by specifying exact coordinates.

**What this means for you:**
You must explicitly specify how widgets should be positioned and sized in their parent window or container.

### Thread-Safe but Single-Threaded
<u>How tkinter thinks about this:</u> Tkinter’s mainloop is single-threaded, meaning it can only handle one task at a time. Long-running tasks will block the GUI, making it unresponsive.

**What this means for you:**
For tasks that take time (e.g., file I/O, API calls), you should use threads or after() to avoid blocking the GUI.

### Events and Event Handlers


You might have noticed the button flashed when you clicked on it​.

  - It generated an “event” which is a notification that the user (or other piece of code) did something.​

Events are a key component of any GUI. They allow the developer to create a mechanism to allow the user to interact with the application in a convenient and intuitive way.
- Before events, users had to remember a plethora of text commands. It limited the power of flexibility of a lot of applications.
- Events work with the keyboard, mouse, touchscreens, voice commands, even the movement of your eye.

Making an event work, means defining the event and the action to take place once the event happens and then binding them to a UI control, like a button.

So let's get the button to "do something" when we click on it, something a little more interesting that just flashing.

We do this by attaching a widget’s event to an event handler​

  - Use either the command attribute while instantiating or the bind() method afterward​.



### The Message Box

Before we go any further, let's take a quick look at a special widget called a messagebox​.

Until now, we have posted information via the **print** function.
- With print, you can post messages to the console or a notebook cell.
- But GUIs don't generally come with a console. So we will need a different approach.

A *message box* is a popup to display messages to the user. You've probably interacted with them countless times over the years.​

Most languages have a “Hello World” app using the messagebox so we’ll create one as well​.
- The "Hello World" program is a staple of programming, usually the first actual program a novice writes in a particular language.
- The difference here is our program uses a window and a button.

Consider the following code to create a button to display a message box​:

```python

# Python code to create a window with a button.
# The button will display a pop-up message when clicked.
import tkinter as tk
from tkinter import messagebox

window = tk.Tk()
# Set the window to an appropriate size. Tkinter isn't very good at figuring out the best size.
# It usually makes the window just big enough so the controls fit.
window.title("Python Message boxes")
window.geometry("300x200")
window.config(bg="#5FB691")

# This is an event-handler. It is a function designed to be called when a particular event occurs.
# Another name for this function is a "callback".
# An event-handler can be as simple or complex as you need.
def msg1():
    messagebox.showinfo("Hello World!", "Hi! Welcome to Python GUIs.")

tk.Button(window, text="Click Me", command=msg1).pack(pady=50)
window.mainloop()
```

Another way to look at a messagebox is as a special kind of window, used to display messages to the user, like a print function.​

Messages can be for anything: help, errors, prompts, whatever you think will assist the user in understanding your app​.

  - Message boxes have a number of forms:​

    - Show infomation, errors, abd warnings to the user.​

    - Ask questions of the user for which answers can be yes/no, ok/cancel, yes/no/cancel and retry/cancel.​
  
Since some message boxes request input from the user, there must be a mechanism to determine what that input is.​

Use one of the following “asks” methods​ for this purpose.

  - **askyesno()** – response is True/False​

  - **askquestion()** - response is “yes”/”no”​

  - **askokcancel()** – response is “ok”/”cancel”​

  - **askretrycancel()** – response is “retry”/”cancel”​

  - **askyesnocancel()** – response is “yes”/”no”/”cancel”​

### Text Widget 

The Text widget is like the Entry widget except it will allow you to input multiple lines of text​.

It uses the same three main operations (get, delete, insert) with some slight variations.

See what a blank window looks like and try typing in some text​:

```python
# Sample text widget in a window.
import tkinter as tk
window = tk.Tk()
window.title("Text Widget")
window.geometry("300x200")
text_box = tk.Text()
text_box.pack()
window.mainloop()
```

### Text get(), delete(), and insert() methods

As mentioned, there is a slight difference between the Text and Entry **get** methods.​
- The latter requires no arguments and returns the entire line.​
- The former requires at a starting point and an optional ending point in this format​.

**\<line>.\<char pos>**

**\<line>.\<char pos>**

The Text widget provides the special parameter tk.END so if you want to get the contents of the entire widget you would try the following:​

```python
# Get all the text in the text widget.
import tkinter as tk
from tkinter import messagebox
window = tk.Tk()
window.title("Text with get()")

def show_selection():
    selection = text_box.get("1.0", "1.5") # first digit is line
    # second digit is character on that line
    messagebox.showinfo("Text get()", "You chose " + selection)

# get everything
def show_everything():
    selection = text_box.get("1.0", tk.END)
    messagebox.showinfo("Text get()", "You chose " + selection)

button = tk.Button(
    text="Get Some!",
    padx=5,
    pady=5,
    bg="blue",
    fg="yellow",
    command = show_selection
)
button.pack(pady=10)
button2 = tk.Button(
    text="Get All!",
    padx=5,
    pady=5,
    bg="blue",
    fg="yellow",
    command = show_everything
)
button2.pack(pady=10)

text_box = tk.Text()
text_box.pack()
window.mainloop()
```

Same goes for delete. Supply the starting and optional ending parameters​:

```python
# Working with the delete method of the Text widget.
import tkinter as tk
from tkinter import messagebox
window = tk.Tk()
window.title("Text with delete()")

# Delete the first five characters in the text widget.
def delete_selection():
    # The two parameters are actually four.
    # The number to the left of the decimal is the line number.
    # The number to the right of the decimal is the character number.
    # The first number is the start, the second number is the "up to" value, not inclusive.
    text_box.delete("1.0", "1.5")
    messagebox.showinfo("Text delete()", "Text deleted")

# get everything
def delete_everything():
    # tk.END tells Python to delete all the way to the end.
    text_box.delete("1.0", tk.END)
    messagebox.showinfo("Text delete()", "All text deleted")

button = tk.Button(
    text="Delete Some!",
    padx=5,
    pady=5,
    bg="blue",
    fg="yellow",
    command = delete_selection
)
button.pack(pady=10)
button2 = tk.Button(
    text="Delete All!",
    padx=5,
    pady=5,
    bg="blue",
    fg="yellow",
    command = delete_everything
)
button2.pack(pady=10)

text_box = tk.Text()
text_box.insert(1.0, "It was the best of times\nIt was the worst of times")
text_box.pack()
window.mainloop()
```

For insert, user the **\<line>.\<column>** format to specify the insertion position​ in the character array.

Insert will either insert or append text​ as appropriate.

```python
import tkinter as tk
from tkinter import messagebox
window = tk.Tk()
window.title("Text with insert()")

# Insert text at the beginning of the widget.
def insert_beginning():
    text_box.insert(1.0, "Hello")

# Insert text at the end of the current array of characters in the widget.
def insert_end():
    text_box.insert(tk.END, "World!")

button = tk.Button(
    text="In front",
    padx=5,
    pady=5,
    bg="blue",
    fg="yellow",
    command = insert_beginning
)
button.pack(pady=10)
button2 = tk.Button(
    text="In back",
    padx=5,
    pady=5,
    bg="blue",
    fg="yellow",
    command = insert_end
)
button2.pack(pady=10)

text_box = tk.Text()
text_box.pack()
window.mainloop()
```

### Frames

Up until now, our UIs have been rather haphazard and don’t look that great, but we can spruce things up a bit using frames.

- A Frame is a widget that can contain and organized other widgets to give your app a more professional look​.

- Do this by “packing” controls inside of a frame rather than the window​ itself.

Adding a frame to a window is just like with any other widget​:

  - Create the frame​

  - Pack the frame​

To assign a widget, you set the widget’s **master** attribute​.

```python
# Adding a label to a frame.
import tkinter as tk

window = tk.Tk()
frame_a = tk.Frame()

label_a = tk.Label(master=frame_a, text="I'm in a frame!")
label_a.pack()

frame_a.pack()
window.mainloop()
```

Now consider two frames, each with a label widget​:

  - Each uses their master to assign to the correct frame​.

  - Frames are packed in the order you choose​.

  - Switching the order, switches the frames​.

```python
# Packing two labels inside of two separate frames.
import tkinter as tk

window = tk.Tk()
frame_1 = tk.Frame()
frame_2 = tk.Frame()

label_1 = tk.Label(master=frame_1, text="I'm in the first frame.")
label_1.pack()

label_2 = tk.Label(master=frame_2, text="I'm in the second frame")
label_2.pack()

frame_1.pack()
frame_2.pack()

window.mainloop()
```

To create a more visually appealing frame, set the relief attribute​.

This creates a border around the frame to give it specified look​, such as:

  - Raised, Sunken, Grooved, Flat or Ridged​

```python
# Creating a frame with a more visually appealing border.
# Start by enumerating the various frame effects:
import tkinter as tk

border_effects = {
    "flat": tk.FLAT,
    "sunken": tk.SUNKEN,
    "raised": tk.RAISED,
    "groove": tk.GROOVE,
    "ridge": tk.RIDGE,
}

window = tk.Tk()

for relief_name, relief in border_effects.items():
    frame = tk.Frame(master=window, relief=relief, borderwidth=5)
    frame.pack(side=tk.LEFT)
    label = tk.Label(master=frame, text=relief_name)
    label.pack()

window.mainloop()
```

### Geometry Managers 

You have already seen how to use **.pack()**. It organizes a widget on a window. It is one of the *geometry managers* Tkinter supports, but there are two other you will find useful:​

- **.place()**

- **.grid()**

By default, **pack()** places frames in the window in the order they are packed.​

Frames are placed in a “parcel” that is as wide as the window and as tall as the frame.​

  - Each frame is placed in the topmost available position​.

Placement occurs vertically unless otherwise specified​.

You get a stacked result like this:

```python
# Showing how to stack frames in a window.
import tkinter as tk

window = tk.Tk()
frame1 = tk.Frame(master=window, width=100, height=100, bg="red")
frame1.pack()

frame2 = tk.Frame(master=window, width=50, height=50, bg="yellow")
frame2.pack()

frame3 = tk.Frame(master=window, width=25, height=25, bg="blue")
frame3.pack()
window.mainloop()
```

Frames can fill a space horizontally by using the fill attribute​:

```python
# Stacking horizontally by using fill.
import tkinter as tk

window = tk.Tk()
frame1 = tk.Frame(master=window, height=100, bg="red")
frame1.pack(fill=tk.X)

frame2 = tk.Frame(master=window, height=50, bg="yellow")
frame2.pack(fill=tk.X)

frame3 = tk.Frame(master=window, height=25, bg="blue")
frame3.pack(fill=tk.X)
window.mainloop()
```

The previous cell fills horizontally, but not vertically. Use the side attribute to specify which side a frame is attached to in order to get a vertical fill​.

```python
# Stacking vertically using fill with side attributes.
import tkinter as tk

window = tk.Tk()
frame1 = tk.Frame(master=window, width=200, height=100, bg="red")
frame1.pack(fill=tk.Y, side=tk.LEFT)

frame2 = tk.Frame(master=window, width=100, bg="yellow")
frame2.pack(fill=tk.Y, side=tk.LEFT)

frame3 = tk.Frame(master=window, width=50, bg="blue")
frame3.pack(fill=tk.Y, side=tk.LEFT)
window.mainloop()
```

Finally, you can allow for resizing in both directions​

  - Set fill to **tk.BOTH**​

  - Set **expand** to True​

```python
# Stacking frames to allow for resizing horizontally and vertically.
import tkinter as tk

window = tk.Tk()

frame1 = tk.Frame(master=window, width=200, height=100, bg="red")
frame1.pack(fill=tk.BOTH, side=tk.LEFT, expand=True)

frame2 = tk.Frame(master=window, width=100, bg="yellow")
frame2.pack(fill=tk.BOTH, side=tk.LEFT, expand=True)

frame3 = tk.Frame(master=window, width=50, bg="blue")
frame3.pack(fill=tk.BOTH, side=tk.LEFT, expand=True)

window.mainloop()
```

Use **place()** to control the precise location of a widget​.

Consider the metric as the number of pixels left and down from the origin (0, 0)​.

The origin is at the top left of the window (or frame)​.

You’ll need to be careful when aligning both the window and the controls, especially if you have a lot of widgets​.

Note the layout doesn’t change​ if you resize the window.

This can be tricky to pull off because OS’s may also affect the look as well as the underlying computing device.​

```python
# Using fixed sizes with the place method.
import tkinter as tk

window = tk.Tk()
frame = tk.Frame(master=window, width=150, height=150)
frame.pack()

label1 = tk.Label(master=frame, text="I'm at (0, 0)", bg="red")
label1.place(x=0, y=0)

label2 = tk.Label(master=frame, text="I'm at (75, 75)", bg="yellow")
label2.place(x=75, y=75)
window.mainloop()
```

**.grid()** gives you the ease of use of **.place()** with the flexibility of **.pack()** so it is often the best choice for your app​
- But it is a little more work​.

You lay out a rectangular surface with a series of “cells”, each of which can hold one or more controls​.

Similar to how you might design web pages.​

Here’s a 3x3 layout of labels​:
- Note the frame is “grided” to the window while the label is “packed” to the frame​

```python
# Code snippet showing how to use a grid to align widgets.
import tkinter as tk

window = tk.Tk()

for i in range(3):
    for j in range(3):
        frame = tk.Frame(
            master=window,
            relief=tk.RAISED,
            borderwidth=1
        )
        frame.grid(row=i, column=j)
        label = tk.Label(master=frame, text=f"Row {i}\nColumn {j}")
        label.pack()

window.mainloop()
```

You can space out the cells in a grid using the **padx**, **pady** attributes:​

```python
# Example on how to apply the padx, pady attributes.
import tkinter as tk

window = tk.Tk()

for i in range(3):
    for j in range(3):
        frame = tk.Frame(
            master=window,
            relief=tk.RAISED,
            borderwidth=1
        )
        frame.grid(row=i, column=j, padx=5, pady=5)
        label = tk.Label(master=frame, text=f"Row {i}\nColumn {j}")
        label.pack()

window.mainloop()
```

You can also apply padding to the internal control that is being “packed”.​

You’ll note, however, resizing isn’t always very pretty.​

```python
# Padding an internal control
import tkinter as tk

window = tk.Tk()

for i in range(3):
    for j in range(3):
        frame = tk.Frame(
            master=window,
            relief=tk.RAISED,
            borderwidth=1
        )
        frame.grid(row=i, column=j, padx=5, pady=5)
        label = tk.Label(master=frame, text=f"Row {i}\nColumn {j}")
        label.pack(padx=5, pady=5)

window.mainloop()
```

We can fix the resizing problem with the columnconfigure and rowconfigure attributes​.

Set the weight attribute to ensure the expansion rate is the same in both directions​.

You can also pad the interior widgets​.

```python
# Using columnconfigure and rowconfigure
import tkinter as tk

window = tk.Tk()

for i in range(3):
    window.columnconfigure(i, weight=1, minsize=75)
    window.rowconfigure(i, weight=1, minsize=50)

    for j in range(0, 3):
        frame = tk.Frame(
            master=window,
            relief=tk.RAISED,
            borderwidth=1
        )
        frame.grid(row=i, column=j, padx=5, pady=5)
        label = tk.Label(master=frame, text=f"Row {i}\nColumn {j}")
        label.pack(padx=5, pady=5)

window.mainloop()
```

Another thing you can do is set the alignment of a widget inside a grid cell by setting the sticky attribute​. The sticky attribute allows the widget to resize according to how the container itself is being resized.

  - “n” or “N” aligns to the top center​

  - “e” or “E” aligns to the right center side​

  - “s” or “S” aligns to the bottom center​

  - “w” or “W” aligns to the left center side​

Combine letters to get a widget to “stick” to multiple sides of a cell​.

To fill an entire cell, set the sticky to “nsew”​ (North, South, East, West):

```python
# Using sticky attributes to auto-resize controls
import tkinter as tk

window = tk.Tk()
window.columnconfigure(0, minsize=250)
window.rowconfigure([0, 1], minsize=100)

label1 = tk.Label(text="A")
label1.grid(row=0, column=0, sticky="ne")

label2 = tk.Label(text="B")
label2.grid(row=1, column=0, sticky="sw")

window.mainloop()
```

### More on Events

Now we have a better understanding of widgets, let’s get back to events​.

The **mainloop()** function we created keeps track of all the events we specify, maintaining a list as we go​.

If an event on the list has a specified event handler, that event handler will get called by the interpreter.​

- It knows to call the event handler if we have specified the **bind** method or the **command** attribute.
    - These are Pythons mechanism to attach widget event handlers to events.​

Let's create a tiny counter, applying a decrease event handler and an increase event handler:​

```python
# Using event handlers.
# In this case apply the command attribute to attach event handler code to an event.
# Note, we haven't explicity specified an event. In that case, Python assumes the default event.
# For the button, it is the click event.
import tkinter as tk
window = tk.Tk()
window.rowconfigure(0, minsize=50, weight=1)
window.columnconfigure([0, 1, 2], minsize=50, weight=1)

def increase():
    value = int(lbl_value["text"])
    lbl_value["text"] = f"{value + 1}"

def decrease():
    value = int(lbl_value["text"])
    lbl_value["text"] = f"{value - 1}"

btn_decrease = tk.Button(master=window, text="-", command=decrease)
btn_decrease.grid(row=0, column=0, sticky="nsew")

lbl_value = tk.Label(master=window, text="0")
lbl_value.grid(row=0, column=1)

btn_increase = tk.Button(master=window, text="+", command=increase)
btn_increase.grid(row=0, column=2, sticky="nsew")

window.mainloop()
```

# Show Your Knowledge

Take everything you've learned and create a simple four-function claculator using Tkinter and Python

> 1. Support +, -, *, /​
>
> 2. User inputs two operands and one operator​
>
> 3. Have an “undo” button to remove the last entry​
>
> 4. Have an “=“ to display the result of the calculation​
>
> 5. Have a “clear” button to reset the calculator back to empty​
>
> 6. Organize your numbers and controls in a nice, clean rectangular format​

```python
# Solution
# clear all variables
objects = dir()

for obj in objects:
  if not obj.startswith("__"):
    del globals()[obj]
    
# our calculator needs 10 numbers, 4 operators, a clear button
# an = button and a decimal button, so make this a 5x4 grid
# with two displays sitting at the top
# one display is for all the user selections
# the other display is for the result of the calculation
# or the current button pressed (corresponding number)
import tkinter as tk

# parts of the calculation
operand1 = ""
operand2 = ""
operator = ""
lastButton = "" # actually this will be Button widget

def parse_calc():
    calc = operand1 + " " + operator + " " + operand2
    return calc

# event handler
def on_btn_click(event, arg):
    global lastButton
    global operator
    global labelDisplay
    lastButton = arg
    btnText = lastButton["text"]
    
    if btnText == "+":
        operator = "+"
        do_operator(operator)
    elif btnText == "-":
        operator = "-"
        do_operator(operator)
    elif btnText == "*":
        operator = "*"
        do_operator(operator)
    elif btnText == "/":
        operator = "/"
        do_operator(operator)
    elif btnText == "U":
        do_undo()
    elif btnText == "C":
        do_clear()
    elif btnText == "=":
        do_equals()
    else:
        do_other(btnText)

    labelDisplay["text"] = parse_calc()
    window.update_idletasks()
    

# helper functions
def do_operator(op):
    global operator
    operator = op
    
def do_undo():
    global operator
    global operand1
    global operand2
    if operand2 != "":
        operand2 == ""
    elif operator != "":
        operator = ""
    elif operand1 != "":
        operand1 = ""
        
            
def do_clear():
    global operator
    global operand1
    global operand2

    operand1 = ""
    operand2 = ""
    operator = ""
    
def do_other(number):
    global operand1
    global operand2    

    if(operator != ""):
        operand2 += number
    else:
        operand1 += number
    
def do_equals():
    global operand1
    global operand2    
    global operator    
    val1 = float(operand1)
    val2 = float(operand2)
    result = 0
    global labelResult
    if operator == "+":
        result = val1 + val2
    elif operator == "-":
        result = val1 - val2
    elif operator == "*":
        result = val1 * val2
    else:
        result = val1 / val2
    labelResult["text"] = str(result)
    operator = ""
    operand1 = ""
    operand2 = ""
    window.update_idletasks()
    
# display
window = tk.Tk()
window.title("My Calculator")
# put the display as label on the first row
frame = tk.Frame(master=window, relief=tk.RAISED, borderwidth=1)
frame.grid(row=0, column=0, columnspan = 4)
labelDisplay = tk.Label(master=frame, background="#34A2FE", text="0")
labelDisplay.pack()
# result label goes on the second row
frame = tk.Frame(master=window, relief=tk.RAISED, borderwidth=1)
frame.grid(row=1, column=0, columnspan = 3)
labelResult = tk.Label(master=frame, background="#FFA2FE", text="0")
labelResult.pack()

# clear button next to result
frame = tk.Frame(master=window, relief=tk.RAISED, borderwidth=1)
frame.grid(row=1, column=2)
button = tk.Button(master=frame, width=2, height=1,
                   text="C")
button.bind("<Button-1>", lambda event
            , arg=button: on_btn_click(event, arg))
button.pack()

frame = tk.Frame(master=window, relief=tk.RAISED, borderwidth=1)
frame.grid(row=1, column=3)
button = tk.Button(master=frame, width=2, height=1,
                   text="+")
button.bind("<Button-1>", lambda event
            , arg=button: on_btn_click(event, arg))
button.pack()

    
for i in range(2, 6):
    for j in range(0, 4):
        frame = tk.Frame(master=window,
            relief=tk.RAISED, borderwidth=1)
        frame.grid(row=i, column=j)
        button = tk.Button(master=frame, width=2, height=1,
                   text="x")

        if(i==5 and j == 1):
            button["text"] = "0"
        elif(i==4 and j == 1):
            button["text"] = "2"
        elif(i==4 and j == 0):
            button["text"] = "1"
        elif(i==4 and j == 2):
            button["text"] = "3"
        elif(i==3 and j == 1):
            button["text"] = "5"
        elif(i==3 and j == 0):
            button["text"] = "4"
        elif(i==3 and j == 2):
            button["text"] = "6"
        elif(i==2 and j == 1):
            button["text"] = "8"
        elif(i==2 and j == 0):
            button["text"] = "7"
        elif(i==2 and j == 2):
            button["text"] = "9"
        elif(i==2 and j == 3):
            button["text"] = "-"
        elif(i==3 and j == 3):
            button["text"] = "*"
        elif(i==4 and j == 3):
            button["text"] = "/"
        elif(i==5 and j == 3):
            button["text"] = "="
        elif(i==5 and j == 2):
            button["text"] = "."
        elif(i==5 and j == 0):
            button["text"] = "U"
        
        if button["text"] != "x":
            button.bind("<Button-1>", lambda event
                        , arg=button: on_btn_click(event, arg))
            button.pack()

window.mainloop()
```
