## Terms
Before we go any further, let's define a few terms we need for this discussion:

### The Class

- A “class” is a design document (in software) we use to create objects.
    - It defines methods, properties, permissions, behaviors and other object-related attributes.
    - Think of it like an architect's blueprint for a house. It *isn't* a house, but houses are built from it.​
    - The class specifies all the behaviors and attributes for the corresponding object.
    - Use the **class** keyword when defining a class
  
  - A **class** contains:
    - Documentation
    - Variables (instance, static, and local)
    - Methods (instance, static, and class)
  


### The Object

An “object” is an “instantiation” (or creation) of a variable based upon a class. We can then use the variable/object in our code.​

- A “method” is a function built into an object, presumably (but not always) to operate on the object itself.
  - Consider a car. "Starting" the car is a method of the car's and works upon the car's engine. You would not expect a car to have a "print taxes" method any more than you would expect your printer to have a "honk horn" method. ​

- A “property” is a state variable built into an object​. In the car analogy it might be something like "running", "idling", "braking."

- An “interface” is a behavior specification for an object​. An interface for a car might be something like a door lock. Lots of things, not just cars, have locks
and they, more or less, work the same way.

- The “constructor” is the method called that does the actual object creation. It tells the program how to build the object. Objects all have a default constructor, but you might want to create your own if you have special requirements for your object's creation.​

<br>

Objects help us to accomplish the goals we discussed earlier:
  - They “hide” or encapsulate complicated procedures behind easier to use methods and properties​.

  - They allow us to, via inheritance, reuse code or extend it​.

  - They allow us to embed functionality into our programs that would be extremely difficult or impossible to do ourselves​.

  - They ensure certain behaviors occur in our programs (through interfaces)​.

  - They allow us to easily package useful functions and methods for general purpose use​.

    - Their use can help reduce errors in our code​.

```python
# Here is an empty class. It doesn't look like much, but behind the scenes there is a lot of code
# to help us create and manage this class.
class Employee:
    ''' This is an employee class '''

# Now that the class definition exists, we can look up information about it.
print(Employee.__doc__)
print("---------M2-----------")
help(Employee)

# If you want to create the class using the Python-provide default constructor.
var1 = Employee()
```

### The Constructor

As mentioned, the constructor is a special type of function called to "instantiate" an object.
- This creates the object from the class definition and passes it back to the caller.
- The default constructor is used when the caller doesn't pass any parameters in the call.

Often, you want to create a class with a particular state​ which is different from the default.
- In this case you might create your own constructor and use that instead.
    - It's not quite the same as "overloading" a method, but it's a similar idea.

Use the constructor to initialize the object upon creation (note the use of **self** here – which we’ll discuss in a bit).​

Use **\__init__** to denote a constructor​.

```python
# This class definition provides a new constructor.
# The default constructor is no longer available, since you provided your own.
# Python doesn't suport function overloading but
# you can "emualate" a default constructor by supplying default values in your custom constructor.
class Employee:
    ''' This is an employee class '''
    def __init__(self, name="Test", age=25, staffnumber=25, designation="Test"):
        self.name = name
        self.age = age
        self.staffnumber = staffnumber
        self.designation = designation
```

### Decorators

The next little Pythonic thing is called a “decorator.”​ Decorators allow us to make more flexible objects.

It allows us to add polymorphism to functions. **Polymorphism** is a term used to describe how software can display different behaviors for the same method call depending upon the type of object that happens to be underneath. For example, a method "make_noise" would meow if the object was a cat and bark if the object was a dog. It's the same make_noise call, but the output is different depending upon ​where it originates.

A decorator is a way to extend existing functions by supplying functions as parameters​, which allows you to "replace" a function with a different one on the fly. It is a kind of "closure". We will discuss closures in more detail a little later.

It's a tricky concept so let’s look at an example. Suppose you want to supply some sort of greeting function:

```python
def greet(myname):
    print("Hi " + myname + "! Welcome to learning decorators!")

greet("Tom")
greet("Latham")
greet("Michael")
```

So far so good. But what if you wanted to change the greeting based upon some input paramters?
- Unlike other languages, Python doesn't support function overloading, so what do you do?

One option is to pass a dictionary (<b>\*\*kwargs</b>) but that can be rather clunky.

We would like something equivalent but maybe a little more explicit.

<br>

The solution is to extend the function using a **decorator** that modifies the output.

```python
# An example decorator.
# There is a function myfunc we want to overload.
# It provides two different variations of myfunc, the original and the new version.
# Which one it decides to use depends upon the name input to the function.
def my_decorator(myfunc):
    # The actual function being passed has a parameter called myname.
    # So what we do is build a new function to "overload" the old one here.
    def func1(my_name):
        # Here is the extension we add to the original call.
        if my_name=="Michael":
            print("Hello " + my_name + "! Your functionality is extended!")
        # Or, we can just call the regular function.
        else:
            myfunc(my_name)
    return func1
```

To get the decorator to work, one more step is required to bind the decorator to the original function you want to extend.

Like this:​

```python
# To bind the decorator above to the method below, use the @ symbol with the decorator name above the function definition.
# Now define the original function.
@my_decorator
def greet(my_name):
    print("Hello " + my_name + "! Welcome to learning decorators!")


greet('Tom')
greet('Latham')
greet('Michael')
```

### Decorator Chaining

Turns out, if you want some real flexibility, Python lets you decorate a decorator.​

When the code runs, it will first execute the inner decorator before running the outer decorator.​

In other words, order matters when you create your decorators. Let's see how this might work:

First we have to figure out the ordering. In this silly example, we want to pretty up some text.
- Our goal is to provide a line of # marks on the outsides of our text and @ marks on the inside and the actual text in the middle.
- The trick is to understand which decorator gets called in which order​.

```python
# This is the outer decorator. Essentially we are bracketing whatever is inside with a bunch of # marks.
def my_hash(func):
    # note the parameters make this function pretty generic
    def my_inner(*args, **kwargs):
        print("#" * 25)
        func(*args, **kwargs)   # note this will call my_at_symbol
        print("#" * 25)
    return my_inner

# This is the inner decorator. Essentially we are bracketing whatever is inside with a bunch of @ marks.
def my_at_symbol(func):
    def inner1(*args, **kwargs):
        print("@" * 25)
        func(*args, **kwargs)   # this will call print
        print("@" * 25)
    return inner1
```

Now look at the binding below. Here is what happens:
- 1. **my_hash** gets called and prints a row of # marks.
- 2. **my_hash** calls the function it "decorates" (passing all the related <b>*args</b> and <b>**kwargs</b>). The decorated function is **my_at_symbol**.
- 3. **my_at_symbol** prints a row of @ marks. It then calls the function it decorates. The decorated function is **my_print**.
- 4. **my_print** prints the text and leaves. Control passes back to the caller, **my_at_symbol**.
- 5. **my_at_symbol** prints another row of @ marks and leaves. Controls passes back to the caller, **my_hash**.
- 6. **my_hash prints** another row of # marks and leaves.


```python
@my_hash
@my_at_symbol
def my_print(mymsg):
    print(mymsg)
my_print("Welcome to decorator chaining")
```

### Class Decorators

What about a class? Can a class use a decorator? Sure!​

The syntax is the same, only decoration is defined inside the class declaration.​

```python
# Decorator we want to use for a class.
def special_name(method):
    def inner(refname):
        if refname.myname == "Rufus":
            print("Sweet! It's Rufus")
        else:
            method(refname)
    return inner

# This is the actual class.
class DisplayName:
    def __init__(self, myname):
        self.myname = myname
    
    # We want to modify this function, so we put the
    # decorator here.
    @special_name
    def my_display(self):
        print(f"My name is: {self.myname}")

# Try any old name.
my_obj1 = DisplayName("Alice")
my_obj1.my_display()

# Try our buddy Rufus.
my_obj1 = DisplayName("Rufus")
my_obj1.my_display()
```

```python
# Solution-We will look at this from two different angles.
# 1. Creating the decorator outside the class.
# 2. Creating the decorator inside the class.

# This decorator is outside the class
def dog_noise(method):
    def inner(refname):
        if refname.breed.lower() == "dachshund" or refname.breed.lower() == "chihuahau":
            print("In dog_yip")
            return "Yip!"
        else:
            print("In dog_bark")
            return method(refname)
    return inner

class Dog:
    ''' Dogs are awesome'''
    # constructor
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    # Our decorator    
    @dog_noise
    def DogBark(self):
        return "Bark!"

Fido = Dog("Fido", "Beagle")
print(f"My dog {Fido.name} is a {Fido.breed} and he will {Fido.DogBark()}")
print()
Fido = Dog("Fido", "Dachshund")
print(f"My dog {Fido.name} is a {Fido.breed} and he will {Fido.DogBark()}")
print()

# Here, the decorator is inside the class.
# It is a slightly different way to create it, generating
# a tighter bind between the class and the decorator.
class Dog:
    # Actual decorator created here.
    def dog_noise(method):
        def inner(self):
            if self.breed.lower() == "dachshund" or self.breed.lower() == "chihuahau":
                print("In dog_yip")
                return "Yip!"
            else:
                print("In dog_bark")
                return method(self)
        return inner

    ''' Dogs are awesome'''
    # Constructor
    def __init__(self, name, breed):
        self.name = name
        self.breed = breed

    # Decorator applied to method here.    
    @dog_noise
    def dog_bark(self):
        return "Bark!"
    

Fido = Dog("Fido", "Doberman")
print(f"My dog {Fido.name} is a {Fido.breed} and he will {Fido.dog_bark()}")
print()
Fido = Dog("Fido", "Chihuahau")
print(f"My dog {Fido.name} is a {Fido.breed} and he will {Fido.dog_bark()}")
```

### Instance Variables.

Now that we have some experience with objects, let's take another look at variables. There are several kinds of variables used in Python classes:
  - Global variables: Most of the variables we've seen fall into this category. They exist during the entire session.

  - Local variables: These are declared inside of methods and only exist in method scope.
  
  - Class variables: These are declared in the class and are shared among objects of the same class.
  
  - Instance variables: These are declared in an object's constructor and a separate copy is created for each object.
  
A common use of an instance variable is for individual class attributes.
 
 - In the constructor, create the instance variable and assign it, if appropriate.​
 
 - Elsewhere, you need to reference it with the keyword **self**.

```python
class DisplayName:
    def __init__(self, my_name):
        # Instance variable
        self.my_name = my_name
    
    # Let's give this class a sample method.
    def my_display(self):
        print(f"My name is: {self.my_name}")


myobj = DisplayName("Kevin")
myobj.my_display()
```

```python
import math

class SqrtPlus1:
    # This "looks" right, but will generate an error, as we shall see.
    def my_sqrt(self, number):
        return 1 + math.sqrt(self.value_to_sqrt)

# Create an instance of SqrtPlus1
sqrt_five = SqrtPlus1()

# Call the instance as if it were a function
result = sqrt_five(5)
print(result)
```

The cell above has code that doesn't quite work right.

```python
import math

class SqrtPlus1:
    # This "looks" right, but will generate an error, as we shall see.
    def __call__ (self, number):
        return 1 + math.sqrt(number)

# Create an instance of SqrtPlus1
sqrt_five = SqrtPlus1()

# Call the instance as if it were a function
print(sqrt_five(5))
print(sqrt_five(10))
```

### Getters and Setters

While not exactly Pythonic, good OOP design uses properties (getters and setters in some languages)​.

- This is a facet of encapsulation. We want to avoid having objects externally access instance variables as a general rule​.

- Instead we give the consumer getters and settings. Here’s how you might use them:​

```python
# A getter than returns an object's name
def get_student(self):
    return self.name

# A setter to manage an object's age
def set_age(self, age):
    self.age = age
```

### Assigning Variables 

Setters, in this case, allow you to control what goes into your objects. Remember the phrase "garbage in, garbage out?"
- Shocking, we know, but users have been known to input bad data from time to time.
- Setters can prevent this from occurring by examining the data prior to inserting it.
    - They can either ignore bad data, alter it, or throw an exception and let the calling application handle the problem.

```python
# Here we are setting the age directly - this is OK.
john.age = 19

# But what if the user decides to input bad data? This works but is probably not a good idea.
john.age = -1
```

In the absence of a setter, in order to prevent bad values from being assigned, you need to add code to handle each assignment​.

```python
# Here some code you might use to test for a valid age.
age = -1
if age <= 0:
    raise ValueError('The age must be positive')
else:   # OK to assign.
    john.age = age
```

The problem is that code could be in multiple places in your application.
- You would need that check every where you made an assignment. That's possibly a lot of extra code.

Better to use a setter and put the code in one place.
- Also better because your object should have the best grasp of its business rules.
- Also this reduces the amount of information a bunch of programmers with plenty of other things to do
to have to remember when working with the object.

Let's create a new class, this time with a getter and setter.​

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.set_age(age) # It's OK to use the setter inside your class too.

    # Setter. Usually it is a good idea to use the prefix "set" for the setter.
    def set_age(self, age):
        if age <= 0:
            raise ValueError('The age must be positive')
        self._age = age

    # Getter. Recommended to use the prefix "get" for the getter.
    def get_age(self):
        return self._age
    
person = Person("Rufus", 30)
print(person.get_age())
```

### Attributes

**Attributes** are like a window to the class's variables. In other languages, they are known as properties and serve as a reference to state information within the class.
- Any time you declare variables in the \__init__ constructor, you create instance attributes which can be referenced using the underscore​ _.

For example, in the class above, **age** is the variable and **_age** is the attribute.​

Consumers don’t have access to the variable directly, which is what we want, but they do have access to the attribute assigned to the variable.​

```python
# Instance of the Person class
john = Person('John', 18)
# You can't access the variable directly because it has an instance attribute.
john.age  # error
```

```python
# You can, however, use the attribute to get at the data.
john._age # OK
```

## Properties

Getters and Setters are prefectly fine tools and have been around from the beginning but many languages are leaning more toward using properties instead.
- Properties provide a kind of software mask that make them look and feel like variables to a consumer, while still providing all the software protection we enjoy from getters and setters.

In Python, properties are considered a slightly better option, but will require a little extra work.​

First you start with some instance variables​.

Consider a Person class​:

```python
# Typical class, two instance variables: name, age
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


john = Person('John', 18)
```

Create a property for a class by adding a property object to the class​

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    # Setter
    def set_age(self, age):
        if age <= 0:
            raise ValueError('The age must be positive')
        self._age = age

    # Getter
    def get_age(self):
        return self._age
    
    # Deleter
    def age(self):
        del self._age

    
    # Create the property by declaring the getter and setter
    age = property(fget=get_age, fset=set_age, fdel=age)
```

Steps to create and configure a property in a class​:
- 1. Create the instance variable in the constructor.
- 2. Create the getter method.
- 3. Create the setter method.
- 4. Create the deleter method.
- 5. Assign the property property(fget=None, fset=None, fdel=None, doc=None).
  - Use fget to set the getter.
  - Use fset to set the setter.
  - Use fdel to assign the deleter.
  - Add a docstring or comment for help.​

All the parameters are optional. So if you want a read-only property, simply set the **fset=None**.​

Usage is now much the same as other languages:

```python
john = Person('John', 18)
print(john.age)
john.age = 20
print(john.age)
# Use the deleter to remove the attribute.
del john.age
print(john.age)
```

### What is the Deleter?

You may have asked this question, looking over the code. The getter and setter are pretty obvious, but why do we need a deleter?

The reason is that over time, as you import and work with large datasets, you start to use up available memory. Deleters allow Python to clear out unused variables in order to free up memory for other uses. For a small class like the Person class, it won't make much of a difference, but you should always be mindful of your memory usage so you don't end up in a situation where memory is unavailable. The **del** command works on global variables and other Python objects so it is nice to also have that option in your own classes.

## Using Decorators for Properties

Python provides a shortcut to the above steps using decorators. It makes the code a little easier to understand. Here's how it works:

```python
# Creating a class with a property using decorators.
class Person:
    def __init__(self, name, age):
        # Note the use of the underscore to denote these are instance variables.
        self._name = name
        self._age = age

    # Decorate the instance variable as the property.
    # Also defines the getter.
    @property
    def age(self):
        return self._age

    # Decorate the setter.
    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value

    # Decorate the deleter
    @age.deleter
    def age(self):
        del self._age

    # This is an output routine for something like a print function.
    # It is a default method and is used by routines that want to provide information
    # on an object without having to know anything about it.
    # In this case the information is the objects name and age.
    def __str__(self):
        return f"My name is: {self._name}, My age is: {self._age}"

```

In lieu of an explicit property assignment where we pass all the functions as arguments, we can specify decorators instead.
The steps are more or less the same except the getter is implied in the **property** decorator and you don't need an explicit assignment.

```python
# Create an instance of Person
person = Person("Rufus McGee", 30)

# Access the age property
print(person.age) 

# Set the age property
person.age = 35
print(person.age)

# Attempt to set a negative age (raising a ValueError)
try:
    person.age = -5
except ValueError as e:
    print(e)

# Get a string representation of the object
print(person)
```

Once a property is deleted, it is cleared from memory and no longer available to the program.

```python
# Delete the age property
del person.age

# Try to access deleted property (raising an AttributeError)
try:
    print(person.age)
except AttributeError as e:
    print(e)
```

## Take note of a subtle difference between the two versions.

Notice a minor variation in the code between the function and decorator versions of the Person class:

The class uses method definitions with standard getter/setter names (**get_age**, **set_age**). While the decorator just uses the property **age** with the method definition. The is because the decorator handles the type of method itself so there is no need to explicity define one.

```python
# Solution using standard assignment function
def dog_noise(method):
    def inner(refname):
        if refname.breed.lower() == "dachshund" or refname.breed.lower() == "chihuahau":
            return "Yip!"
        else:
            return method(refname)
    return inner

class Dog:
    ''' Dogs are awesome'''
    # constructor
    def __init__(self, name, breed, age):
        self._name = name
        self._breed = breed
        self.set_age(age)
        
    @dog_noise
    def dog_bark(self):
        print("In dog_bark")
        return "Bark!"
    
    # Setter
    def set_age(self, age):
        if age <= 0:
            raise ValueError('The age must be positive')
        self._age = age
    
    # Getter
    def get_age(self):
        return self._age
    
    # Deleter
    def age(self):
        del self._age
    
    # Assign everything
    age = property(fget=get_age, fset=set_age, fdel=age)
```

```python
# Testing basic functions
Fido = Dog("Fido", "Beagle", 3)
print(f"My dog {Fido._name} is a {Fido._breed} and he is {Fido.age} years old")
# Should give us an error
Fido.age = -4
```

```python
# This is OK
Fido.age = 4
print(f"My dog {Fido._name} is a {Fido._breed} and he is {Fido.age} years old")
del Fido.age
# This should throw an error.
print(Fido.age)
```

```python
# Solution using decorators
def dog_noise(method):
    def inner(refname):
        if refname.breed.lower() == "dachshund" or refname.breed.lower() == "chihuahau":
            return "Yip!"
        else:
            return method(refname)
    return inner

class Dog:
    ''' Dogs are awesome'''
    # constructor
    def __init__(self, name, breed, age):
        self._name = name
        self._breed = breed
        self._age = age
        
    @DogNoise
    def dog_bark(self):
        print("In DogBark")
        return "Bark!"
    
    # Getter
    @property
    def age(self):
        return self._age
        
    # Setter
    @age.setter
    def age(self, age):
        if age <= 0:
            raise ValueError('The age must be positive')
        self._age = age

    # Deleter
    @age.deleter
    def age(self):
        del self._age
```

```python
# Testing basic functions
Fido = Dog("Fido", "Beagle", 3)
print(f"My dog {Fido._name} is a {Fido._breed} and he is {Fido.age} years old")
# Should give us an error
Fido.age = -4
```

## Nested Functions

Let's go back to functions for a minute. Did you know that Python allows you to "nest" function definitions?
- Some compiled languages will allow you to nest classes, but not functions directly.
<br>

It's an unusual construct, so let's look at an example and then discuss why you might want to do this.

```python
# An example of a nested function.
def outside_fn(str1):
    # We're actually defining a function inside another function.
    def inside_fn():
        print(str1 + " Inner Function")
    # Here we run our own code.
    print(str1 + " Outer Function")
    # Then we call the function we defined.
    inside_fn()

outside_fn("Hello ")
```

There are actually a number of benefits from creating and using nested functions:
<ul>
    <li>1. Encapsulation 
        <p>Nested functions can encapsulate specific functionality and variables within an inner function. This helps reduce the number of global variables and avoids name clashes. This is particularly useful for utility functions that are only needed within the context of the enclosing function.<p>
    </li>
    <li>2. Organization
        <p>Grouping related functionality together through nested functions can make the code more organized and readable. This can be useful in coding complex functions where distinct tasks are performed that can be isolated into distinct subroutines. They can also work in the same way as traditional functions. If a piece of code is repeated, then turning it into a function means not having to copy/paste the same code in multiple places.</p>
    </li>
    <li>3. Closures
        <p>We will explore closures shortly. Closures are useful for factories, callbacks and decorators because they can maintain state after the outermost function has finished executing.</p>
    </li>
    <li>4. Functional Programming
        <p>In functional programming, functions are nothing more than variables that can be passed into and out of other functions. This is a common technique in vectorizing operations and allows for more flexible and reusable code.</p>
    </li>
    <li>5. Lazy Evaluation
        <p>A nested functions can be set aside until it is explictly needed. This can help with performance optimization when the exact timing of execution is important.</p>
    </li>
</ul>

### Closures

A *closure* is a function that defines a nested function and passes that function to a caller.

Closures can improve the functionality and flexibility of your code in a number of ways:
- Decorators are a kind of closure. We have already seen closures used as decorators ealier in this section.
- Closures are used to generate callback functions. Callbacks are very common for event-driven programming, used in graphical user interfaces.
- Closures can serve as the basis for a "factory", which is a function that generates other functions. Factories are helpful when you need to repeat
function code with only a slight alteration based on some parameter:

```python
# Here's a simple example of a function factory
# It's a simple "rooter" that takes a number n and creates a function
# That will generate the nth root of an input.
# Hence, we can create a whole slew of root functions on the fly.
def root_factory(root_value):
    def root_function(number):
        return number ** (1/root_value)
    return root_function

# Create specific functions using the factory, square root, cube root, fourth root
square_root = root_factory(2)
cube_root = root_factory(3)
fourth_root = root_factory(4)

# Test the functions
print(square_root(25))
print(cube_root(27))
print(fourth_root(256))
```

## Functions as Parameters

Because functions are an object like any other variable, Python allows us to pass them just as we would any other varaible.
- We have already seen functions used as paramters in closures, but it is even simpler than that. You can just create functions and pass them to any other function:

```python
# Create a couple of simple functions.
def myUpper(mystr):
    return mystr.upper()

def myLower(mystr):
    return mystr.lower()

# New create a function that takes as a parameter another function.
# This is a pretty common practice in data science. We often want to 
# create data-dependent functions during the feature engineering process
# to handle data from independent sources or data that varies widely under
# certain conditions.
# An example would be sales data of say, beachware, in summer and winter seasons.
def worldVirus(myfunc):
    virus = myfunc("Covid-19")
    print(virus)

# Test
worldVirus(myUpper)
worldVirus(myLower)
```
