### Radio button

Radio buttons are handled by the widget **radiobutton**.

Use a radiobutton whenever you want to present the user with mutually exclusive choice between options​.

Each button has text and a value​

  - Start by creating a dictionary to make it easier​.

    -The key is the text, the value is the value​.

```python
from tkinter import *
from tkinter.ttk import *

# Creating master Tkinter window
master = Tk()
master.geometry("175x175")

# Tkinter string variable
# able to store any string value
v = StringVar(master, "1")

# Dictionary to create multiple buttons
values = {"RadioButton 1" : "1",
		"RadioButton 2" : "2",
		"RadioButton 3" : "3",
		"RadioButton 4" : "4",
		"RadioButton 5" : "5"}

# Loop is used to create multiple Radiobuttons
# rather than creating each button separately
for (text, value) in values.items():
	Radiobutton(master, text = text, variable = v,
		value = value).pack(side = TOP, ipady = 5)

# Infinite loop can be terminated by
# keyboard or mouse interrupt
# or by any predefined function (destroy())
mainloop()
```

### Setting Events

As before, there are two ways to add events to the radio button (like with any other widget)​:

- Create a method and bind to the widget​ using the **command** attribute.

- Create an event handler and bind to the event​ using the **bind()** method.

  - This method is more flexible if you aren’t using the default events​.

  - The bind works like this: widget.bind(event, handler, add=None)​.

Note: The **add** parameter (default value = False) tells Python whether to add this handler to an already existing handler(s) or replace the existing handler. In rare cases you will want to bind multiple functions to the same event.

First let's look at the command attribute:

```python
# Simple example of binding using the command attribute.
import tkinter as tk
from tkinter import messagebox

# Function to be called when the button is clicked
def on_button_click():
    messagebox.showinfo("Button was clicked!", "In the Event Handler!")

# Create the main window
root = tk.Tk()
root.title("Binding Using command Attribute")

# Create a button and bind it to the on_button_click function using the command attribute
button = tk.Button(root, text="Click Me!", command=on_button_click)
button.pack(pady=20)

# Run the main event loop
root.mainloop()
```

Next, the **bind** method. Note some subtle but important differences between the two approaches:

```python
# Simple example of binding using the bind method.
import tkinter as tk
from tkinter import messagebox

# Function to be called when the button is clicked
# Note here, that we are passing an event parameter. That object will have details about the actual event we can use to pass information from the control to the handler.
# The event object is customizable so we can add whatever information we think is appropriate.
def on_button_click(event):
    messagebox.showinfo("Button was clicked!", "In the Event Handler!")

# Create the main window
root = tk.Tk()
root.title("Binding Using bind Method")

# Create a button
button = tk.Button(root, text="Click Me")
button.pack(pady=20)

# Bind the button's left-click event to the on_button_click function
# Note here we specify the event we want to handle.
button.bind("<Button-1>", on_button_click, add=None)

# Run the main event loop
root.mainloop()
```

### Widget Arrays

If you are going to bind to an array of radio buttons, you will want to make use of the **StringVar**​.

The **StringVar** is a window-based variable used to hold the value of a control (a string)​.

**IntVar** is the integer version of this variable​.

For an array of controls, like radio buttons, set the text and value for each widget independently​.
- Then reference StringVar or IntVar to know specifically which control sent the event.

Dictionaries make this pretty straightforward​.

```python
# Add events to the radio button - using the command option first.
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox

window = Tk()
window.title("Python Radio Buttons")
window.geometry('175x175')
window.config(bg="#5FB691")

# Declare the StringVar. We will use it to figure out which radio button is active.
v = StringVar(window, "1")

def msg1():
    print("Hello World! With widget binding!!" + v.get())

# Style class to add style to Radiobutton
# it can be used to style any ttk widget
style = Style(window)
style.configure("TRadiobutton", background = "light green",
				foreground = "red", font = ("arial", 10, "bold"))

# Dictionary to create multiple buttons
values = {"RadioButton 1" : "1",
		"RadioButton 2" : "2",
		"RadioButton 3" : "3",
		"RadioButton 4" : "4",
		"RadioButton 5" : "5"}

# Loop is used to create multiple Radiobuttons
# rather than creating each button separately
for (text, value) in values.items():
	Radiobutton(window, text = text, value = value, variable = v,
				command=msg1).pack(side = TOP, ipady = 5)

# Infinite loop can be terminated by
# keyboard or mouse interrupt
# or by any predefined function (destroy())
mainloop()
```

Next, let's use the **bind()** method

```python
# Bind the event handler to the widget using bind.
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox

window = Tk()
window.title("Python Radio Buttons")
window.geometry('175x175')
window.config(bg="#5FB691")

# Tkinter string variable
# able to store any string value
v = StringVar(window, "1")
    
def handle_click(event):
    print("Hello World! With event binding!!" + v.get())
    print("X value is: " + str(event.x))
    print("Y value is: " + str(event.y))
    # iterate through all the widgets on the window
    for child in window.children.items():
        if(type(child[1]) is Radiobutton):
            rb = child[1]
            print("It's a radio button")
            print(rb["value"])


# Style class to add style to Radiobutton
# it can be used to style any ttk widget
style = Style(window)
style.configure("TRadiobutton", background = "light green",
				foreground = "red", font = ("arial", 10, "bold"))

# Dictionary to create multiple buttons
values = {"RadioButton 1" : "1",
		"RadioButton 2" : "2",
		"RadioButton 3" : "3",
		"RadioButton 4" : "4",
		"RadioButton 5" : "5"}

# Loop is used to create multiple Radiobuttons
# rather than creating each button separately
for (text, value) in values.items():
    rb = Radiobutton(window, text = text, variable = v,
				value = value)
    # there's a problem here in that the variable is changed
    # after the event is called. We'll have to fix that.
    rb.bind("<Button-1>", handle_click)
    rb.pack(side = TOP, ipady = 5)

mainloop()
```

However, there’s a problem with this approach​, as noted in the code.
- Turns out the variable changes AFTER the event happens so we need to try something different to get it right.​

The **\<Button-1>** event denotes a mouse click on the left most button (button 1)​

So to make this work, we need to capture an event that occurs after this one​, rather than this event.

Turns out we have just such an event: the mouse-up event, denoted by **\<ButtonRelease-1>**

Let's see how it works:

```python
# Fixing the button click problem, using the button release event.
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox

window = Tk()
window.title("Python Radio Buttons")
window.geometry('175x175')
window.config(bg="#5FB691")

# Tkinter string variable
# able to store any string value
v = StringVar(window, "1")
    
def handle_click(event):
    print("Hello World! With event binding!!" + v.get())
    print("X value is: " + str(event.x))
    print("Y value is: " + str(event.y))


# Style class to add style to Radiobutton
# it can be used to style any ttk widget
style = Style(window)
style.configure("TRadiobutton", background = "light green",
				foreground = "red", font = ("arial", 10, "bold"))

# Dictionary to create multiple buttons
values = {"RadioButton 1" : "1",
		"RadioButton 2" : "2",
		"RadioButton 3" : "3",
		"RadioButton 4" : "4",
		"RadioButton 5" : "5"}

# Loop is used to create multiple Radiobuttons
# rather than creating each button separately
for (text, value) in values.items():
    rb = Radiobutton(window, text = text, variable = v,
				value = value)
    # The click event fires too early, so use the button release event instead.
    rb.bind("<ButtonRelease-1>", handle_click)
    rb.pack(side = TOP, ipady = 5)

mainloop()
```

### Variable Trace 

Turns out there is yet another way to fix the problem with our radio buttons, but it requires the trace method (for the StringVar/IntVar)​

The trace method allows us to attack a callback method to a variable so we can track when it changes​

  - Similar to an event handler, but for a value instead​

```python
# Sample of how to trace variable change state, note this will not run.
v.trace("w", lambda name, index, mode, var=v: callback)
```

```python
# Using trace in a script.
from tkinter import *

window = Tk()
window.title("Python Radio Buttons with Trace")

# Function which will be run every time StringVar is changed
def callback(*args):
    selected_value = radio_var.get()
    print(f"Value changed to {selected_value}")
    

# Create variable to hold radio button value
radio_var = StringVar(window, "1")

# Tracing variable change state
radio_var.trace("w", callback)

values = {"RadioButton 1" : "1",
        "RadioButton 2" : "2"}

for (text, value) in values.items():
    rb = Radiobutton(window, text = text, variable = radio_var, value = value)
    rb.pack(side = TOP, ipady = 5)

mainloop()
```

### Menus

Use menus for common functions that handle the entire app, such as opening/closing or saving a file or exiting the program.​

Menus can also serve as the container of other widgets​ (like a button).

Menus sit on the “menubar” at the top of the application​ or as a free-floating menu, or context/popup menu (enabled by right-clicking).

Add a standard menu to an application with the following steps:​

  1. Create a menubar
  2. Add a toplevel menu (that has no commands)
  3. Add menu commands (specifying the label (text) and the associated command)​
  4. Add other toplevel menus and repeat
  5. Add the menubar to the window​
  
The first step is creating the menubar, which is a hidden widget that holds the menus​.

- “Cascade” visible menus off the menubar​

Associate each menu with a command method (or command) which runs when the menu is selected​.

```python
from tkinter import *
from tkinter.ttk import *
from tkinter import messagebox


def mnu_new_file():
    messagebox.showinfo("New file menu", "New file menu selected" )
def mnu_open_file():
    messagebox.showinfo("Open file menu", "Open file menu selected" )
def mnu_save_file():
    messagebox.showinfo("Save file menu", "Save file menu selected" )

# creating tkinter window
window = Tk()
window.title("Using Menus in Python")

# Step 1: Creating Menubar
menubar = Menu(window)

# Step 2: Adding File Menu toplevel to the menubar
file = Menu(menubar, tearoff = 0)  # top-level menu for the app
# set the menus to cascade off the file menu
menubar.add_cascade(label ="File", menu = file)

# Step 3: add the menu commands  
file.add_command(label ="New File", command = mnu_new_file)
file.add_command(label ="Open File", command = mnu_open_file)
file.add_command(label ="Save File", command = mnu_save_file)
file.add_separator() # this creates a line for grouping
# notice in this next line, we just add a single command
file.add_command(label ='Exit', command = window.destroy)

# Adding Help Menu
help_ = Menu(menubar, tearoff = 0)
menubar.add_cascade(label ='Help', menu = help_)
help_.add_command(label ='Tk Help', command = None)
help_.add_command(label ='Demo', command = None)
help_.add_separator()
help_.add_command(label ='About Tk', command = None)

# Step 4: add the display Menu to the main window
window.config(menu = menubar)
mainloop()
```

### Popup Menus

Also known as context menus, popup menus are designed to specific help centered around the part of the application that has focus.
- Unlike regular menus, which are available at all times and are generic to the application.

To setup a popup menu, you need two primary parameters​

  - Master or parent widget​

  - A set of options which is a comma-separated list or key-value pairs to set menu properties, such as​

    - Background/foreground colors​

    - Cursor, highlight color, image, padding and many others​

Set the menu variable for reference.​

```python
# Python example using tkinter for a popup menu.
import tkinter
			
window = tkinter.Tk()
window.geometry('300x300')
frame1 = tkinter.Label(window, width = 400,
								height = 400, bg = '#AAAAAA')
frame1.pack()

def hey(s):
    frame1.configure(text = s)

#create menu
popup_menu = tkinter.Menu(window, tearoff = 0)
popup_menu.add_command(label = "say hi", command = lambda:hey("hi"))
popup_menu.add_command(label = "say hello", command = lambda:hey("hello"))
popup_menu.add_separator()
popup_menu.add_command(label = "say bye", command = lambda:hey("bye"))

#display menu on right click
def do_popup(event):
	try:
		popup_menu.tk_popup(event.x_root, event.y_root)
	finally:
		popup_menu.grab_release()

window.bind("<Button-3>", do_popup)
tkinter.mainloop()
```

You might have noticed the use of the **lambda** here​.

The command property expects a method, but what if you want to pass a parameter? You can’t, technically.​

Lambdas allow us to specify both the method AND one or more parameters to use​.

### Progress Bar 

Use the progress bar to indicate completion status for some operation​.

- Users like to know something is *happening*​ and that the application hasn't frozen.

Set the progress bar’s value, then run the window’s **update_idletasks** method to refresh the bar’s appearance.​

Progress bars can be oriented horizontally or vertically and be determinate or indeterminate.
- Indeterminate is where you don’t show the entire bar.​

```python
# Example progress bar with Python/tkinter in determinate mode.
from tkinter import * 
from tkinter.ttk import *
import time

# creating tkinter window
window = Tk()

# Progress bar widget
progress = Progressbar(window, orient = HORIZONTAL,
			length = 100, mode = 'determinate')

def show_bar():
    barValue = 0
    while barValue <= 100:    
        progress['value'] = barValue
        window.update_idletasks()
        time.sleep(.01)
        barValue += 1

progress.pack(pady = 10)

# This button will initialize the progress bar
Button(window, text = 'Start', command = show_bar).pack(pady = 10)

# infinite loop
mainloop()
```

### Working with the listbox

Add a listbox to your application with the following steps:

1. Create the listbox.
2. Add one or more items​

Each item is a two-part object​

  - Part 1 is an index value which organizes the items​

  - Part 2 is the actual value to get when selected​

Listboxes can be set for mutually exclusive choices, such as a military rank, or they can be set for combinations of choices, such as the types of movies one might enjoy. 

<div style="font-size:18px" align=right>Expert</div>

```python
# Python/tkinter example of working with a listbox.
# This is a multi-selection listbox, so choices are not mutually exclusive.
import tkinter as tk
from tkinter import ttk, messagebox

# Show choices
def show_selected():
    selected_indices = listbox.curselection()
    selected_items = [listbox.get(i) for i in selected_indices]
    # Output to console and Message box for reference.
    print("Selected items:", selected_items)
    messagebox.showinfo("Selected Items", "You have selected: " + ", ".join(selected_items))

# Create the main window.
root = tk.Tk()
root.title("Multi-Select Listbox")

# Create a label.
label = ttk.Label(root, text="Select your favorite fruits:")
label.pack(pady=10)

# Create a listbox with multiple selection mode.
# Note the selectmode flag.
listbox = tk.Listbox(root, selectmode=tk.MULTIPLE, height=10)
fruits = ["Apple", "Banana", "Cherry", "Date", "Grapes", "Mango", "Orange", "Peach", "Strawberry", "Watermelon"]

for fruit in fruits:
    listbox.insert(tk.END, fruit)

listbox.pack(pady=10)

# Create a button to show selected items.
button = ttk.Button(root, text="Show Selected", command=show_selected)
button.pack(pady=10)

# Run the application.
root.mainloop()
```

Each listbox has an enumeration of values called a **curselection**​.

Use a for loop or a comprehension to retrieve all the selected items in the list.​

Listboxes are a little different from many other controls in that they house a collection of value​s.
- The values can be objects, so your listbox can be as complex as you like.

Generally, to initialize a listbox, create a list of items using a Tkinter variable object, then assign the list of items to the listbox​.

Once items are assinged, you will probably want to add an event handler to a listbox, maybe to automatically update a selection list.
- Do this by creating an event handler and bind to the listbox default event **ListboxSelect**.​

```python
# Using Python and tkinter to handle listbox collection assignments and events.
import tkinter as tk
from tkinter import ttk
from tkinter.messagebox import showinfo

# create the root window
window = tk.Tk()
window.title('Listbox Selections and Events')
window.geometry('340x200')

# create a list box
langs = ('Java', 'C#', 'C', 'C++', 'Python',
         'Go', 'JavaScript', 'PHP', 'Swift')

varList = tk.Variable(value=langs)

listbox = tk.Listbox(window,
    listvariable=varList, height=6,
    selectmode=tk.EXTENDED
)

listbox.pack(expand=True, fill=tk.BOTH)

def items_selected(event):
    # get all selected indices
    selected_indices = listbox.curselection()
    # get selected items
    selected_langs = ",".join([listbox.get(i) for i in selected_indices])
    msg = f'You selected: {selected_langs}'
    print(msg)


listbox.bind('<<ListboxSelect>>', items_selected)

window.mainloop()
```

### Opening New Windows

Something else you might need to do is open additional windows​ to show different aspects or functions in your application.

- Do this by using **Toplevel**. It creates a new window and binds it to the main application window.​

```python
# Create an application and add a second window.
from tkinter import *
from tkinter.ttk import *

# creates a Tk() object
master = Tk()

# sets the geometry of main
# root window
master.geometry("200x200")

# Function to open a new window
# on a button click.
def open_new_window():
	
	# Toplevel object which will
	# be treated as a new window
	new_window = Toplevel(master)

	# sets the title of the
	# Toplevel widget
	new_window.title("New Window")

	# sets the geometry of toplevel
	new_window.geometry("200x200")

	# A Label widget to show in toplevel
	Label(new_window,
		text ="This is a new window").pack()

label = Label(master,
			text ="This is the main window")

label.pack(pady = 10)

# a button widget which will open a
# new window on button click
btn = Button(master,
			text ="Click to open a new window",
			command = open_new_window)
btn.pack(pady = 10)

# mainloop, runs infinitely
mainloop()
```

### Dialog Windows

Use dialogs to handle common tasks.

Tkinter provides some standard dialogs most applications use. For example:

  - Open a file for use​.

  - Save an open file​.

  - Color chooser​.

To use a dialog, import it and configure as necessary​.

- For example, use the open file dialog to allow users to choose a file on the file system.​

  - Import the askopenfile class from the filedialog module.​

  - Set the file filters as appropriate.​

```python
# Using the Open file dialog
from tkinter import *
from tkinter.ttk import *

# importing askopenfile function
# from class filedialog
from tkinter.filedialog import askopenfile

window = Tk()
window.geometry('200x100')

# This function will be used to open
# file in read mode and only Python files
# will be opened
def open_file():
    file = askopenfile(mode ='r', filetypes =[('Python Files', '*.py')]) 
    file.close()
    print(file.name)
    
btn = Button(window, text ='Open', command = lambda:open_file())
btn.pack(side = TOP, pady = 10)

mainloop()
```

Use the save file dialog (asksaveasfile) to allow users to choose a name for an output file​.

There are many properties in this dialog you can customize for a specific look​:

```python
# Using the save file dialog.
from tkinter import *
from tkinter import ttk

# import only asksaveasfile from filedialog
# which is used to save file in any extension
from tkinter.filedialog import asksaveasfile

root = Tk()
root.geometry('200x150')

# function to call when user press
# the save button, a filedialog will
# open and ask to save file
def save():
	files = [('All Files', '*.*'),
			('Python Files', '*.py'),
			('Text Document', '*.txt')]
	file = asksaveasfile(filetypes = files
                      , defaultextension = files)

btn = ttk.Button(root, text = 'Save', command = lambda : save())
btn.pack(side = TOP, pady = 20)

mainloop()
```

Use the color chooser to pick a color from a color palette​:

```python
# Example Python/tkinter using a color chooser.
from tkinter import *
# importing the choosecolor package
from tkinter import colorchooser

# Function that will be invoked when the
# button is clicked in the main window.
def choose_color():
	# Variable to store hexadecimal code of color
	color_code = colorchooser.askcolor(title ="Choose color")
	print(color_code)

window = Tk()
button = Button(window, text = "Select color",
				command = choose_color)
button.pack()
window.geometry("300x300")
window.mainloop()
```
