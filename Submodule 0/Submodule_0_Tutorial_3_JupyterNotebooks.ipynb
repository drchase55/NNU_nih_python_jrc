## Introduction to Jupyter Notebooks

**What is a Jupyter Notebook?**

It is a tool that lets you <u>write text</u> and <u>run computer code</u> (typically R or Python) all mixed into a single file, making it easy to see and understand. That is, Instead of just writing programming code in one tool and the descriptions or explanations in another file or a textbook, it blends them together sequentially.
<br>
Also, Jupyter lets you type that computing code into boxes (called "cells") and run each part separately. 
<br>
As it turns out, this page was written as a Jupyter Notebook but as the user, you only see it as a web page. Since it doesn't have any code (yet) to run, it basically is a web page.
<br>
BUT, when you look at a Jupyter notebook in the right kind of interface, the user can edit the text and the code, not just view it.
<br>
<br>
<i>**and that user can change the code and run it!**</i>

With Jupyter Notebook, you can also add notes, explanations, and even pictures to help make sense of your work. This makes it a great tool for learning, experimenting, and sharing code with others.

Think of it like a notebook where you can write instructions (code), run them step by step, and take notes—all in one place!

# Why Use Jupyter for Coding and Data Science?

Jupyter Notebooks are useful because:

1. It’s Easy to Use – You don’t need to run an entire program at once. You can run small pieces of code step by step, see what happens, and fix mistakes as you go.
2. You Can Take Notes While You Code – Instead of keeping your notes in a separate document, you can write explanations right next to your code to make it easier to understand later.
3. It Helps You See Your Results Clearly – If your code creates a table, chart, or graph, Jupyter Notebook will show it right inside the document.
4. It's Great for Learning and Sharing – You can save your notebook and send it to someone else so they can see your work and try it themselves.

That is, Jupyter Notebooks are especially useful for students, researchers, and anyone working with data because it helps break things into small, easy-to-understand steps.



# Running Code in Jupyter Notebooks

To run a single code cell:

- Click inside the cell to select it.
- Press Shift + Enter on your keyboard or click the Run button (right-pointing triangle for 'play' in the toolbar)

The code will execute, and any output (like text, numbers, or graphs) will appear below the cell.
