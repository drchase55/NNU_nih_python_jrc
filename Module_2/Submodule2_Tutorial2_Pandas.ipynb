Vectorization in Pandas can handle a wide variety of situations​

You can apply functions to Pandas objects in three ways:​

  - Table-wise using **pipe()**

  - Row or column-wise using **apply()**

  - Element-wise using **applymap()**
  
Use table-wise functions to operate on an entire dataframe​

In the sample below, we add a number to each element in the table (in this case, 2)

```python
import pandas as pd
import numpy as np

def adder(ele1,ele2):
    return ele1+ele2

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.pipe(adder,2)
print(df.apply(np.mean))
```

By row-wise or column-wise, what we mean is applying a function along a particular axis​

Which means we can use this in higher-order matrices too​

Each axis has an index. List the index to set the application​

Columns is index 0 and default​

Rows is index 1​

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
df.apply(np.mean)
print(df.apply(np.mean))
```

Doing an element-wise operation is the same as our original notion of vectorization​

Note in the sample here we are combining a table-wise and element-wise operation as a comparison​

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(5,3),columns=['col1','col2','col3'])
print(df.apply(np.mean))    # old mean
print(df.applymap(lambda x:x*100)) # modify elements
print(df.applymap(lambda x:x*100).apply(np.mean)) # new mean
```

Iterating over a series (1d) object is similar to an array or tuple​

Iterating over a dataframe or panel is a little trickier​

The iteration for them provides a dictionary object which has the columns/values of rows/values​

  - Use iteritems() for column-wise​

  - Use iterrows() for row-wise​

  - User itertuples() to convert rows to tuples​
  
The **iteritems()** method produces key, value pairs with the column name and the column values as a series​

```python
import pandas as pd
import numpy as np
 
df = pd.DataFrame(np.random.randn(4,3),columns=['col1','col2','col3'])
for key,value in df.iteritems():
    print(key,value)
```

The **iterrows()** method returns a dictionary with each key, value pair as a row index and a series of row values​

Each value is associated with its specific column label​

```python
import pandas as pd
import numpy as np

df = pd.DataFrame(np.random.randn(4,3),columns = ['col1','col2','col3'])
for row_index,row in df.iterrows():
    print(row_index, row)
```

<div style="font-size:18px">Group-wise Operation</div>

The first step is to establish the groups (SPLIT)​

In Pandas, we can use the **.groupby()** method for this​

Use states, and group on the variable we just created ​

```python
# Group By
grp = states.groupby("Population_cat")
grp.mean()
grp.var()
 
# Only One variable
states.Income.groupby(states.Population_cat).mean()
```

We can index the group-by with a dynamically created value

```python
# Group By Dynamic Variables
states.Illiteracy.groupby(states.Income < 4100).mean()
```

We can iterate through the groups, allowing more flexibility to do things

```python
# Group By Iteration
# we need to convert the series 
# to a bunch of # strings so we can grab the first letter
states['firstlet']= states.State.astype(str).str[0]
grp = states.groupby("firstlet")
for (name,data) in grp:
    print(name)
    print(data)
```

<div style="font-size:18px">Aggregation</div> 

Use **grp.agg()** to pass in user written function​

**groupedobj.agg(function)**​

You can also pass in a list of length two tuples with names and functions to name the columns. ​

```python
# Aggregation
# Write your own function
def sterr(x):
    try:
        return np.std(x) / np.sqrt(len(x))
    except:
        return np.nan
sterr(states.Income)
grp = states.groupby(states.Income < 4100)["Illiteracy"]
grp.agg(sterr)
grp.agg([(“spread",sterr),
	   (“center",np.mean)])
```

<div style="font-size:18px">Use Vars instead of Index</div>

We can request the group by object returns variables (columns) instead of index or hierarchical indexes

```python
# Eliminate the index
grp = states.groupby(states.Income < 4100,as_index=False)
grp.mean()
```

<div style="font-size:18px">Joining Groupwise Aggregations</div> 

After calculating groups means, we may wish to rejoin them to the group. ​

Pandas calls this “transform”, where the transformed data is placed back in the context of the original data. ​

The result must be a scalar, or the appropriate sized array. 

```python
# Transform 
grp = states.groupby(states.Income < 4100,as_index=False)["Illiteracy"]

grp.transform(np.mean) 
```

<div style="font-size:18px">Windowing</div> 

What is “windowing?” Imagine you are a car salesman and you want to know if your sales trends are heading in the right direction​

  - Month over month sales might not give an accurate picture as sales can vary quite a bit from month to month​

  - What you need is a “window” of time to smooth out the monthly volatility so you can see if the sales for the window is increasing or decreasing​

Windows consist of several types​

  - Rolling (same size over time)​

  - Expanding (accumulates over time)​

  - Weighted (gives more importance to subsets of the data)​
  
Detailed use and explanation of these functions is beyond the scope of this course, but really it’s a matter of supplying parameters such as the size of the window, how many periods you want and the statistical measures you want to apply​

```python
df = pd.DataFrame(np.random.randn(10, 4),
   index = pd.date_range('1/1/2000', periods=10),
   columns = ['A', 'B', 'C', 'D'])
# rolling window of 3 periods
print(df.rolling(window=3).mean())
# expanding window, miniumum 3 periods
print(df.expanding(min_periods=3).mean())
# weighted average, with decay specified
print(df.ewm(com=0.5).mean())
```
