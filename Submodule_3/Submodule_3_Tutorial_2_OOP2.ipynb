### Instance Variables (Advanced)

Now that we have some experience with objects, let's take another look at variables. There are several kinds of variables used in Python classes:
  - Global variables: Most of the variables we've seen fall into this category. They exist during the entire session.

  - Local variables: These are declared *inside of methods* and only exist in method scope.
  
  - Class variables: These are declared in the *class* and are shared among objects of the same class.
  
  - Instance variables: These are declared in an object's constructor and a separate copy is created for each object.
  
A common use of an instance variable is for individual class attributes.
 
 - In the constructor, create the instance variable and assign it, if appropriate.​
 
 - Elsewhere, you need to reference it with the keyword **self**.
<br>
In the next box, you can see an example of the fact that "translated codon" is not a global variable, so it cannot be accessed outside of the class, as well as examples of other variable types.

```python
# Global variable: Exists outside any class or method and can be accessed anywhere
global_genetic_code = {"ATG": "Methionine", "TAA": "Stop", "TAG": "Stop", "TGA": "Stop"}

class DNAAnalyzer:
    # Class attribute: Shared across all instances of this class
    species_name = "Unknown Species"

    def __init__(self, sequence):
        # Instance attribute: Unique to each object created from this class
        self.instance_sequence = sequence

    def find_start_codon(self):
        # Method-local variables: Exist only within this method during execution
        local_start_codon = self.instance_sequence[:3]
        translated_codon = global_genetic_code.get(local_start_codon, "No start codon found")

        return translated_codon

# Example usage:
sample = DNAAnalyzer("ATGCGTACG")  # Create an instance with a DNA sequence

# Accessing different variable types
print(f"Global variable (genetic code dictionary): {global_genetic_code}")  # ✅ Accessible
print(f"Class attribute (shared across instances): {DNAAnalyzer.species_name}")  # ✅ Accessible
print(f"Instance attribute (unique per object): {sample.instance_sequence}")  # ✅ Accessible
print(f"Method-local variable output: {sample.find_start_codon()}")  # ✅ Accessible via method

# 🚨 Attempting to access a method-local variable directly (WILL CAUSE AN ERROR)
try:
    print(f"Trying to access 'translated_codon' directly: {sample.translated_codon}")  # ❌ Not possible
except AttributeError:
    print("Error: 'translated_codon' is a method-local variable and cannot be accessed outside the method!")
```

## Properties

Getters and Setters are prefectly fine tools and have been around from the beginning but many languages are leaning more toward using properties instead.
- Properties provide a kind of software mask that make them look and feel like variables to a consumer, while still providing all the software protection we enjoy from getters and setters.

In Python, properties are considered a slightly better option, but will require a little extra work.​

First you start with some instance variables​.

Consider a Patient class​:

```python
# Typical class, two instance variables: name, age
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age


john = Person('John', 18)
```

Create a property for a class by adding a property object to the class​

```python
class Patient:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    # Setter
    def set_age(self, age):
        if age <= 0:
            raise ValueError('The age must be positive')
        self._age = age

    # Getter
    def get_age(self):
        return self._age
    
    # Deleter
    def age(self):
        del self._age

    
    # Create the property by declaring the getter and setter
    age = property(fget=get_age, fset=set_age, fdel=age)
```

Steps to create and configure a property in a class​:
- 1. Create the instance variable in the constructor.
- 2. Create the getter method.
- 3. Create the setter method.
- 4. Create the deleter method.
- 5. Assign the property property(fget=None, fset=None, fdel=None, doc=None).
  - Use fget to set the getter.
  - Use fset to set the setter.
  - Use fdel to assign the deleter.
  - Add a docstring or comment for help.​

All the parameters are optional. So if you want a read-only property, simply set the **fset=None**.​

Usage is now much the same as other languages:

```python
john = Person('John', 18)
print(john.age)
john.age = 20
print(john.age)
# Use the deleter to remove the attribute.
del john.age
print(john.age)
```

### What is the Deleter?

You may have asked this question, looking over the code. The getter and setter are pretty obvious, but why do we need a deleter?

The reason is that over time, as you import and work with large datasets, you start to use up available memory. Deleters allow Python to clear out unused variables in order to free up memory for other uses. For a small class like the Person class, it won't make much of a difference, but you should always be mindful of your memory usage so you don't end up in a situation where memory is unavailable. The **del** command works on global variables and other Python objects so it is nice to also have that option in your own classes.

## Using Decorators for Properties

Python provides a shortcut to the above steps using decorators. It makes the code a little easier to understand. Here's how it works:

```python
# Creating a class with a property using decorators.
class Person:
    def __init__(self, name, age):
        # Note the use of the underscore to denote these are instance variables.
        self._name = name
        self._age = age

    # Decorate the instance variable as the property.
    # Also defines the getter.
    @property
    def age(self):
        return self._age

    # Decorate the setter.
    @age.setter
    def age(self, value):
        if value < 0:
            raise ValueError("Age cannot be negative")
        self._age = value

    # Decorate the deleter
    @age.deleter
    def age(self):
        del self._age

    # This is an output routine for something like a print function.
    # It is a default method and is used by routines that want to provide information
    # on an object without having to know anything about it.
    # In this case the information is the objects name and age.
    def __str__(self):
        return f"My name is: {self._name}, My age is: {self._age}"

```

In lieu of an explicit property assignment where we pass all the functions as arguments, we can specify decorators instead.
The steps are more or less the same except the getter is implied in the **property** decorator and you don't need an explicit assignment.

```python
# Create an instance of Person
person = Person("Rufus McGee", 30)

# Access the age property
print(person.age) 

# Set the age property
person.age = 35
print(person.age)

# Attempt to set a negative age (raising a ValueError)
try:
    person.age = -5
except ValueError as e:
    print(e)

# Get a string representation of the object
print(person)
```

Once a property is deleted, it is cleared from memory and no longer available to the program.

```python
# Delete the age property
del person.age

# Try to access deleted property (raising an AttributeError)
try:
    print(person.age)
except AttributeError as e:
    print(e)
```

## Take note of a subtle difference between the two versions.

Notice a minor variation in the code between the function and decorator versions of the Person class:

The class uses method definitions with standard getter/setter names (**get_age**, **set_age**). While the decorator just uses the property **age** with the method definition. The is because the decorator handles the type of method itself so there is no need to explicity define one.

## Nested Functions

Let's go back to functions for a minute. Did you know that Python allows you to "nest" function definitions?
- Some compiled languages will allow you to nest classes, but not functions directly.
<br>

It's an unusual construct, so let's look at an example and then discuss why you might want to do this.

```python
# An example of a nested function.
def outside_fn(str1):
    # We're actually defining a function inside another function.
    def inside_fn():
        print(str1 + " Inner Function")
    # Here we run our own code.
    print(str1 + " Outer Function")
    # Then we call the function we defined.
    inside_fn()

outside_fn("Hello ")
```

There are actually a number of benefits from creating and using nested functions:
<ul>
    <li>1. Encapsulation 
        <p>Nested functions can encapsulate specific functionality and variables within an inner function. This helps reduce the number of global variables and avoids name clashes. This is particularly useful for utility functions that are only needed within the context of the enclosing function.<p>
    </li>
    <li>2. Organization
        <p>Grouping related functionality together through nested functions can make the code more organized and readable. This can be useful in coding complex functions where distinct tasks are performed that can be isolated into distinct subroutines. They can also work in the same way as traditional functions. If a piece of code is repeated, then turning it into a function means not having to copy/paste the same code in multiple places.</p>
    </li>
    <li>3. Closures
        <p>We will explore closures shortly. Closures are useful for factories, callbacks and decorators because they can maintain state after the outermost function has finished executing.</p>
    </li>
    <li>4. Functional Programming
        <p>In functional programming, functions are nothing more than variables that can be passed into and out of other functions. This is a common technique in vectorizing operations and allows for more flexible and reusable code.</p>
    </li>
    <li>5. Lazy Evaluation
        <p>A nested functions can be set aside until it is explictly needed. This can help with performance optimization when the exact timing of execution is important.</p>
    </li>
</ul>

### Closures

A *closure* is a function that defines a nested function and passes that function to a caller.

Closures can improve the functionality and flexibility of your code in a number of ways:
- Decorators are a kind of closure. We have already seen closures used as decorators ealier in this section.
- Closures are used to generate callback functions. Callbacks are very common for event-driven programming, used in graphical user interfaces.
- Closures can serve as the basis for a "factory", which is a function that generates other functions. Factories are helpful when you need to repeat
function code with only a slight alteration based on some parameter:

```python
# Here's a simple example of a function factory
# It's a simple "rooter" that takes a number n and creates a function
# That will generate the nth root of an input.
# Hence, we can create a whole slew of root functions on the fly.
def root_factory(root_value):
    def root_function(number):
        return number ** (1/root_value)
    return root_function

# Create specific functions using the factory, square root, cube root, fourth root
square_root = root_factory(2)
cube_root = root_factory(3)
fourth_root = root_factory(4)

# Test the functions
print(square_root(25))
print(cube_root(27))
print(fourth_root(256))
```

## Functions as Parameters

Because functions are an object like any other variable, Python allows us to pass them just as we would any other varaible.
- We have already seen functions used as paramters in closures, but it is even simpler than that. You can just create functions and pass them to any other function:

```python
# Create a couple of simple functions.
def myUpper(mystr):
    return mystr.upper()

def myLower(mystr):
    return mystr.lower()

# New create a function that takes as a parameter another function.
# This is a pretty common practice in data science. We often want to 
# create data-dependent functions during the feature engineering process
# to handle data from independent sources or data that varies widely under
# certain conditions.
# An example would be sales data of say, beachware, in summer and winter seasons.
def worldVirus(myfunc):
    virus = myfunc("Covid-19")
    print(virus)

# Test
worldVirus(myUpper)
worldVirus(myLower)
```
