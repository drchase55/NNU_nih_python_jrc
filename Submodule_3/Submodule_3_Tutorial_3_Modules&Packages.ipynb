# What is a Module?

We need ways to organize our code
  - For portability

  - For reusability

  - For ease of maintenance

  - For organization



A *module* is simply a mechanism to do all those things​

At its most basic, it is just a file that contains variables, functions and classes you can import into other applications​.

## Python Code Organization

Python organizes code in the following manner:

- A group of statements in a code block is called a **function** (or method)

- A group of functions (and variables and classes in a file) is called a **module**​

- A group of modules (in a folder) is called a **package**

- A group of packages (in one or more folders) is called a **library**
<br>
<br>

As your Python projects get larger and more complex (trust us - they will) you will need to organize your code so that it is easier to test, modify and deploy.

## The &#95;name&#95; Variable 

Up to now, we’ve really just run a bunch of scripts, but often Python programs are often more complicated than that​.

Python uses a variable called \__name__ which is different depending upon how its containing script executes​.

With the double underscores, it’s called a “dunder name”. Dunder stands for double underscores.​

Using \__name__ you can test for an explicit starting point of an application​.
    - This is useful if you are calling scripts from elsewhere and need to know whether a given piece of code is from the main block or is being called from elsewhere.

If you start on the active script, the \__name__ is “\__main__”​.
- Otherwise the \__name__ will be the name of the module, such as **MyModule**.

If you thought this sounded a lot like the c-style “main” you would be correct​.

```python
def myfunc():
    print("Inside function")

if __name__ == "__main__":
    print("Program Execution is direct")
    print("The value of __name__ is", __name__)
    myfunc()


else:
    print("Program execution is indirectly as a module from some other program")
    print("The value of __name__ is", __name__)
```

The issue is often your typical Python script without “main” runs from top to bottom, but often it is preferable to “start” somewhere other than the top.
- Another issue is modules often have test or execution code you don't want to run.​
- This helps to organize different modules to run the way you want.​

```python
# Here's how you might want to use _name_.
def myfunc():
    print("Inside function")

# If this code is being called from somewhere else, _name_ will be different
# so this piece will not run.
if __name__ == "__main__":
    print("Program Execution is direct")
    print("The value of __name__ is", __name__)
    myfunc()

else:
    print("Program execution is indirectly as a module from some other program")
    print("The value of __name__ is", __name__)
```

## Logging 

Being able to track events as they occur is a useful and often necessary function in a program​.
- It greatly helps with debugging and auditing and provides a historical record of what happened without you having to watch it.​
- It’s also a very common requirement​ in production code as well as automated code.

Python, naturally, provides a module to help you do just that​.

**logging** is a package designed to be easy to use and pretty functional and can adapt to virtually any requirement.​

You can log in a number of different places with a number of different levels​.

It can handle most situations, is used by a lot of third-party packages and is thread-safe​.

A simple example lets you post a warning to the screen​:

```python
import logging
logging.warning("Watch out!")
```

Some sample logging code might look like this:

```python
# Set filename to have the logger write to an external file
# Writes can be added and timestamped.
import logging

logging.basicConfig(filename="mywarninglog.txt",level=logging.WARNING)
print("Displaying Warning level demo: ")
logging.debug("Debug message")
logging.info("Info message")
logging.warning("Warning message")
logging.error("Error message")
logging.critical("Critical message")
```

## Capturing the Trace

Often when you get an error, you will want to see the **call stack**​.
- The call stack contains function/method calls, along with their parameters.
- The interpreter and operating system manage the call stack, which maintains program state and tells the CPU what code to run.

The **state** of the program at the point of error is called the *stack trace* and provides function calls and variable values currently on the stack​.

Set the **exc_info** parameter to True​ to display the stack trace in your log. Here are a couple of examples:

```python
# Without setting the exc_info flag to True
# You just get an error.
import logging

num1 = 3
num2 = 0
try:
  divres = num1 / num2
except Exception as e:
  logging.error("Displaying Exception")
```

```python
# Now set the exc_info flag to True
# and you will see the stack trace.
import logging

num1 = 3
num2 = 0
try:
  divres = num1 / num2
except Exception as e:
  logging.error("Displaying Exception", exc_info=True)
```

### Assert

It’s a common practice to use print statements to print variable information to the console while debugging.
- You've seen a lot of it here.
- The problem is at some point you have to remove those print statements when putting code into production.
  
- But then you have to add them again when making changes! Rats!
  

A way to avoid this is to use assertions

  - Assertions only trigger when a particular condition (usually an error) is met.
  
  - So you can leave them in your production code.

```python
def mySum(my_elements):
    # This assertion tests whether the parameter passed is empty.
    assert len(my_elements) != 0, "Empty list"
    return sum(my_elements)

# OK
list1 = [1,2,3,4,5,6]
print("Sum of list1 is:",mySum(list1))

# Error
list2 = []
print("Sum of list2 is:",mySum(list2))
```
