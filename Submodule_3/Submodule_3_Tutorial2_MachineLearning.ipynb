## Over/Under Fitting

![Overfitting and Underfit of data](https://docs.aws.amazon.com/images/machine-learning/latest/dg/images/mlconcepts_image5.png)

There are some general challenges with any modeling. 
Underfitting is when the model is too simple to properly fit the training data. You can see in the image that the model (line) does not capture the relationship between the independent and dependent variables, especially at the ends. 

Overfitting is when the model too closely predicts the training data. In the fit above, the data has some scatter from error, so we want to make a prediction about the general trend. Our model (the line) doesn't really capture what the relationship is between the independent and the dependent variables. Thus, it won't generalize well to new data

The ideal model doesn’t overfit or underfit-- as shown in the center panel.

This is where cross validation is useful

## Cross validation

## Supervised vs Unsupervised learning

Machine learning algorithms can train on data where the predicted values are known
This is supervised learning
E.g. predicting house prices in Boston, purchasers of computers at Best Buy, likely democrats vs republicans, etc.

When there is no “label” machines can find common attributes among groups
This is unsupervised learning
E.g. determining the typical shopper at a specialty store

## Classification vs. Regression

Classification involves separating observations into **groups**
Hey wait, that sounds like clustering, yeah, it sort of is, except that our “clusters” are actual labels defined beforehand
- Promoter vs. non-promoter
- alpha helix vs. beta sheet
- Any one of various risk categories (Low, medium, high)

With classification there are a finite number of possible outcomes

With regression we are predicting a **value-based** result instead of a **class**

- How much will my clunker sell for online?
- How long will a customer put up with crappy service before they bolt?
- How many snow-cones can I expect to sell in the spring vs summer months?
- How many people will vote in the next election?

Some models support both regression and classification while other only support one or the other 

## Clustering

Clustering is an unsupervised technique
It works by separating observations into “groups”
Hey that sounds like classification. Unlike classification, the clusters aren’t defined beforehand. That’s why clustering is **unsupervised.**

Each group has a “center”. Observations belong to what center they are “nearest” to.
Nearest has many definitions, Euclidean distance, radial distance, Manhattan distance, etc.

## Training and Test Data

**Training data** is a critical component in the process of building predictive models using machine learning. It refers to the dataset used to "teach" a model to make predictions or classifications by learning patterns, relationships, and features within the data.

In the context of bioinformatics, training data often comprises biological datasets that include labeled examples (input-output pairs) or unlabeled examples (unsupervised learning). The ultimate goal is to <u>enable the model to generalize from the training data to make accurate predictions on unseen biological data.</u>


You will use the majority of your data set as **training data** for model generation.
Most of your data will be of this type
It is common to use 70% or more of your data to train the model
(You may also want to set aside some data for cross-validation)
Use **test data** to validate your model
You should expect to see a drop off in performance against your test data, but it shouldn’t be significant
For best results, use “newer” data

Use Python modules such as train_test_split (from sklearn.model_selection)

## Assessing the Model Score

There are various metrics we have for assessing how well the model fits the data
Sum of squares regression (SSR)
The difference between the predicted value and the mean
Sum of squares error (SSE)
The observed difference between the observed value and the predicted value
Sum of squares total (SST)
The difference between observed value and the mean
SST = SSR + SSE

There are many metrics for assessing how well a model works
Accuracy (predictions right vs total predictions)
Precision (true positives vs (true pos + false pos)
Recall (true positives vs (true pos + false neg)
F1 (based on both precision and recall)
Cohen Kappa (agreement between categorical items)
…
Each looks at the performance in a slightly different way

# Machine Learning Algorithms

## Random Forest

Random Forest is an ensemble technique using decision trees
Trees are initialized randomly then each “votes” on the winner
One of the most flexible and forgiving machine learning algorithms (and most popular)

## Neural Network

Based upon biological neurons, neural networks are collections of software neurons working in concert to determine an outcome
Neurons take one or more inputs (plus a bias term) and pass them through an activation function

## Support Vector Machine

Support Vector Machines separate classes using a hyperplane
Seek to find the largest distance between classes

## Ensembles

# How do you choose the best model?

Scoring Multiple Models
We can look at individual scores or use something like ANOVA to determine which models are more effective
We also can look at things like
- Generalizability
- Ease of understanding
- Use of features
- Resource usage (how hard to train and run)
- False discovery rates

Regularization – L1

In some datasets the dimensionality is large with respect to the number of samples
Select a small number of features to predict the trend
L1 regularization uses a Laplace distribution and gradient descent to reduce many of the weights of features to zero, 
leaving only the most important features

Outliers

Suppose you had a set of points like this
Most of the points fit rather nicely on a regression line
However, the outliers “pull” the regression line away from the rest of the points
This is another example of “overfitting”

What we need is a way to reduce the impact of outliers on our model training

Regularization – L2

L2 (or ridge) regularization works by penalizing points by the square of the residual
The farther away a point is from the guess, the greater (squared) the penalty
We solve for the weights or coefficients using calculus
The effect is that outliers have much less effect on determining the overall regression line than points closer to the model

This works for many types of models

Model Tuning

Most models allow for hyperparameters which you can tune
This usually requires a good understanding of the internal algorithm
How many layers in a neural network
How many trees in a random forest
Kernel properties in a support vector machine

Train the model, then tune hyperparameters to get that little bit of extra accuracy
