```python
# import the uci Molecular Biology (Promoter Gene Sequences) Data Set
url = 'https://archive.ics.uci.edu/ml/machine-learning-databases/molecular-biology/promoter-gene-sequences/promoters.data'
names = ['Class', 'id', 'Sequence']
data = pd.read_csv(url, names = names)
data.head()
```

A **promoter** is a sequence of DNA which allows the DNA to be transcribed. It is the site where RNA polymerase binds, the enzyme which is responsible for synthesizing the RNA. 

The promoter data set was designed for machine learning on what makes a good promoter sequence. Thus, the Class series can be interpreted as follows
* \+ means the sequence can serve as a promoter
* \- means that the sequence is anything that is **NOT** a promoter
