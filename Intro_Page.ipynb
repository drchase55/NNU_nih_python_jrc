# Introduction to Python for Bioinformatics

## Overview
The primary goal of this project is to help <b>you</b> learn to explore and analyze bioinformatics data using Python. To do this, you must first learn how to use Python itself.

This course will approach the task of learning Python in several stages:

* Programming basics to get you comfortable with how programs are built and run
* Exploring important concepts, such as object-oriented design to help with building user-friendly and useful applications
* Incorporating external libraries to take advantage of tools that have already been written
  - A key tool is [Biopython](https://
* Examining and reengineering your data
* Applying data science and machine learning to biological data sets to derive knowledge.

Previous programming knowledge is HELPFUL, but not required.
  - However, plan on doing a lot of programming in this course

## Learning Objectives
+ You will confidently create and run basic Python programs
+ You will be able to access data sources, particularly LARGE file sources, to make data-based conclusions
    - You will perform data manipulation with NumPy and Pandas
+ You will be able to use AI and statistical measures to draw deep inferences from large biological data sets and to make or evaluate testable hypotheses
+ You will be able to make a GUI of a Python application
